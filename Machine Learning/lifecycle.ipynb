## Machine Learning Lifecycle
    -
        Data Gathering
           ↓
        Data Storage (DataFrame)
           ↓
        Data Exploration
           ↓
        Data Cleaning
           ↓
        Exploratory Data Analysis (EDA)
           (Pandas | Matplotlib | Seaborn)
           ↓
        Data Preprocessing
           ↓
        Machine Learning Model Building


- 1️⃣ Data Gathering
    - Collect data from CSV, APIs, databases, etc.

- 2️⃣ Data Storage (DataFrame)
    - Store data using Pandas DataFrame
    - Makes data easy to manipulate
- 3️⃣ Data Exploration
    - Understand structure
    - Check columns, types, basic stats

- 4️⃣ Data Cleaning
    - Handle missing values
    - Remove duplicates
    - Fix incorrect data

- 5️⃣ Exploratory Data Analysis (EDA)
    - Visualize data using:
        - Pandas
        - Matplotlib
        - Seaborn
    - Find patterns & relationships

- 6️⃣ Data Preprocessing
    - Feature scaling
    - Encoding categorical data
    - Splitting data (train/test)
- 7️⃣ Machine Learning Models
    - Train models (Regression, Classification, etc.)
    - Evaluate performance
