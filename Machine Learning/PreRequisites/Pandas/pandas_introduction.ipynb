# Introduction to Pandas in Python

Pandas is one of the most popular Python libraries for working with structured data. It is widely used in data analysis, data cleaning, and machine learning workflows because it makes tabular data easy to read, transform, and summarize.

## Why use Pandas?

- It provides fast and flexible data structures like `Series` and `DataFrame`.
- It helps you load data from CSV, Excel, SQL, and many other sources.
- It makes filtering, grouping, sorting, and aggregating data simple.
- It is built on top of NumPy and works well with the Python data ecosystem.


## Core Pandas objects

- `Series`: a one-dimensional labeled array.
- `DataFrame`: a two-dimensional labeled table with rows and columns.
- Indexing: used to label and access data efficiently.


```python
import pandas as pd

# A simple Series
marks = pd.Series([85, 90, 78, 92], name='Marks')
print('Series:')
print(marks)

# A simple DataFrame
students = pd.DataFrame({
    'Name': ['Asha', 'Ravi', 'Neha', 'Karan'],
    'Age': [20, 21, 22, 20],
    'Marks': [85, 90, 78, 92]
})

print('\nDataFrame:')
print(students)

# Basic summary
print('\nAverage marks:', students['Marks'].mean())
print('\nStudents scoring above 80:')
print(students[students['Marks'] > 80])
```

## Common use cases

- Cleaning missing or messy data.
- Selecting and transforming columns.
- Joining and merging datasets.
- Creating summaries and reports.
