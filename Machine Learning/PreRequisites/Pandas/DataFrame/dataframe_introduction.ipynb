## DataFrame
- A DataFrame in Python is a 2-dimensional, table-like data structure—think of it like an Excel sheet or a SQL table, but inside your code.
- A DataFrame has:
    - Rows → records (like entries)
    - Columns → fields (like name, age, etc.)
    - Index → row labels

- 🔹 Key Features
    - Structured like a table (rows & columns)
    - Can hold different data types in each column
    - Supports powerful operations like:
        - Filtering
        - Grouping
        - Aggregation
        - Joining (like SQL)


- 🔹 Real-Life Analogy
    - Excel sheet → DataFrame
    - Database table → DataFrame
    - CSV file → easily converted into DataFrame

- 🔹 Common Use Cases
    - Data analysis
    - Machine learning preprocessing
    - Reading files (CSV, Excel, JSON)
    - Reporting and dashboards

### 1. Create DataFrame from Dictionary

    - This method uses a dictionary where keys are column names and values are lists of data.

```python
import pandas as pd

data = {
    'Name': ['Vipin', 'Rahul', 'Amit'],
    'Age': [25, 30, 28],
    'City': ['Delhi', 'Mumbai', 'Pune']
}

df = pd.DataFrame(data)
df
```

### 2. Create DataFrame from List of Dictionaries

    - Each dictionary represents a row in the DataFrame.

```python
data = [
    {'Name': 'Vipin', 'Age': 25, 'City': 'Delhi'},
    {'Name': 'Rahul', 'Age': 30, 'City': 'Mumbai'},
    {'Name': 'Amit', 'Age': 28, 'City': 'Pune'}
]

df = pd.DataFrame(data)
df
```

### 3. Create DataFrame from List of Lists

    - Data is provided as rows (list of lists), and column names are specified separately.

```python
data = [
    ['Vipin', 25, 'Delhi'],
    ['Rahul', 30, 'Mumbai'],
    ['Amit', 28, 'Pune']
]

columns = ['Name', 'Age', 'City']

df = pd.DataFrame(data, columns=columns)
df
```
