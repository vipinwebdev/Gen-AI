## Data Cleaning
    - Before data cleaning , we must have the idea that if the data is having any missing value or not or any duplicate value or not.
        - For missing value check use : isNull()
            - Syntax:
                df.isNull() => Shows boolean in dataframe.
                df.isNull().sum() => for total missing value count for each column.

        - For duplicate values check: duplicated()


```python
import pandas as pd

df = pd.read_csv("student_depression_dataset.csv")
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

## Mathematical Functions in DataFrame

1. `sum()`    ⇒ Calculates the sum of all values present in a column or in multiple columns.
2. `mean()`   ⇒ Calculates the average of all values or mean of all values in a column.
3. `min()`    ⇒ Returns minimum value in a column.
4. `max()`    ⇒ Return maximum value in a column.
5. `median()` ⇒ Calculates the median value of a column.
6. `var()`    ⇒ Calculates the variance of all the values in a column.
7. `axis`     ⇒ It helps us calculate the mathematical calculation row-wise or column-wise.
              `row-wise      ⇒   axis = 1`
              `column-wise   ⇒   axis = 0`
