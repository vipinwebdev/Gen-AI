## Attributes and Functions of DataFrame

A pandas `DataFrame` is a 2-dimensional table with rows and columns.

This notebook uses `student_depression_dataset.csv` so the study notes and examples are based on a real dataset.


```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 120)

df = pd.read_csv('student_depression_dataset.csv')
print(df.head())
```

### 1. What a DataFrame is

- A DataFrame is the main tabular data structure in pandas.
- Rows represent observations and columns represent features.
- In this dataset, each row describes one student record.


```python
print('Rows and columns of the dataset:')
print(df.shape)
print('\nFirst row index:')
print(df.index[0])
print('\nColumn names:')
print(df.columns.tolist())
```

### 2. Important DataFrame attributes

Attributes describe the structure of the DataFrame and do not need parentheses.

- `shape` gives the number of rows and columns.
- `size` gives the total number of cells.
- `ndim` gives the number of dimensions.
- `columns` gives the column labels.
- `index` gives the row labels.
- `dtypes` gives the data type of every column.
- `values` returns the underlying array values.
- `empty` checks whether the DataFrame has no rows.
- `axes` returns both the row index and column labels.
- `T` gives the transpose of the DataFrame.


```python
print('shape:', df.shape)
print('size:', df.size)
print('ndim:', df.ndim)
print('empty:', df.empty)
print('axes:', df.axes)
print('\ndtypes:')
print(df.dtypes)
print('\nvalues preview:')
print(df.values[:2])
print('\ntranspose preview:')
print(df.T.head())
```

### 3. Preview functions

These functions help inspect the dataset quickly.

- `head()` shows the first rows.
- `tail()` shows the last rows.
- `sample()` gives random rows for a quick check.


```python
print('First 5 rows:')
print(df.head())

print('\nLast 5 rows:')
print(df.tail())

print('\nRandom 3 rows:')
print(df.sample(3, random_state=42))
```

### 4. Summary functions

- `info()` shows column names, non-null counts, and dtypes.
- `describe()` gives summary statistics for numeric columns.
- `describe(include='all')` includes both numeric and categorical columns.


```python
print('Dataset info:')
df.info()

print('\nNumeric summary:')
print(df.describe())

print('\nComplete summary:')
print(df.describe(include='all'))
```

### 5. Counting and unique values

- `count()` counts non-null values in each column.
- `nunique()` counts unique values in each column.
- `unique()` returns the distinct values of a specific column.
- `value_counts()` shows the frequency of each category.

These are useful for understanding categorical columns such as `Gender`, `City`, and `Profession`.


```python
print('Non-null count:')
print(df.count())

print('\nUnique counts:')
print(df.nunique())

print('\nUnique cities:')
print(df['City'].unique())

print('\nGender frequency:')
print(df['Gender'].value_counts())
```

### 6. Accessing columns and rows

When column names contain spaces, use bracket notation.

- `df['column']` selects one column.
- `loc[]` selects rows and columns by label.
- `iloc[]` selects rows and columns by position.


```python
print("City and CGPA:")
print(df[['City', 'CGPA']].head())

print('\nFirst 3 rows using loc:')
print(df.loc[:2, ['Gender', 'Age', 'Depression']])

print('\nFirst 3 rows using iloc:')
print(df.iloc[:3, :4])
```

### 7. Filtering data

Filtering is one of the most important DataFrame operations.

- Boolean conditions return only matching rows.
- Use this when you want to study a subset of the dataset.


```python
print('Students with depression = 1:')
print(df[df['Depression'] == 1].head())

print('\nStudents with high academic pressure:')
print(df[df['Academic Pressure'] >= 4].head())

print('\nStudents sleeping 5-6 hours:')
print(df[df['Sleep Duration'] == "'5-6 hours'"].head())
```

### 8. Sorting functions

- `sort_values()` sorts by one or more columns.
- `sort_index()` sorts by the row index.

Sorting is useful for finding highest values, lowest values, or ordered records.


```python
print('Top 5 students by CGPA:')
print(df.sort_values(by='CGPA', ascending=False).head())

print('\nTop 5 students by Financial Stress:')
print(df.sort_values(by='Financial Stress', ascending=False).head())
```

### 9. Missing-value handling

- `isnull()` checks for missing values.
- `notnull()` checks for valid values.
- `dropna()` removes rows with missing values.
- `fillna()` replaces missing values with a chosen value.

If a dataset has no missing values, we can create a small demo copy for learning the method.


```python
print('Missing values per column:')
print(df.isnull().sum())

demo_df = df.copy()
demo_df.loc[demo_df.index[:3], 'CGPA'] = None
demo_df.loc[demo_df.index[:2], 'City'] = None

print('\nDemo data with missing values:')
print(demo_df[['City', 'CGPA']].head())

print('\nAfter fillna:')
print(demo_df.fillna({'City': 'Unknown', 'CGPA': 0})[['City', 'CGPA']].head())

print('\nAfter dropna:')
print(demo_df.dropna(subset=['City', 'CGPA'])[['City', 'CGPA']].head())
```

### 10. Renaming and replacing values

- `rename()` changes column labels or index labels.
- `replace()` changes values in a column or across the whole DataFrame.

These are common cleanup operations before analysis.


```python
renamed_df = df.rename(columns={'Depression': 'Depression_Status', 'Work/Study Hours': 'Study_Hours'})
print('Renamed columns:')
print(renamed_df.columns.tolist())

print('\nReplace values in Gender:')
print(df['Gender'].replace({'Male': 'M', 'Female': 'F'}).head())
```

### 11. Type conversion

- `astype()` changes the data type of a column.
- `to_numeric()` is useful when a column should be numeric but is stored as text.

For learning, we convert a numeric column to a different type and also clean a copied column.


```python
print('Original CGPA dtype:', df['CGPA'].dtype)
print('Converted CGPA dtype:', df['CGPA'].astype('float64').dtype)

cgpa_as_text = df['CGPA'].astype(str)
print('\nCGPA as text sample:')
print(cgpa_as_text.head())
```

### 12. Grouping and aggregation

- `groupby()` splits the data into groups.
- Aggregation functions such as `mean()`, `sum()`, and `count()` summarize each group.

This is useful for comparing patterns across categories.


```python
print('Average depression by gender:')
print(df.groupby('Gender')['Depression'].mean())

print('\nAverage CGPA by profession:')
print(df.groupby('Profession')['CGPA'].mean().head())

print('\nCount of students by city:')
print(df.groupby('City')['id'].count().sort_values(ascending=False).head())
```

### 13. Useful study takeaway

- Attributes tell you what the DataFrame looks like.
- Functions help you inspect, filter, sort, clean, and summarize the data.
- Use bracket notation for columns with spaces or special characters.
- Use `student_depression_dataset.csv` as the base dataset for all examples in this notebook.


```python
print('Final quick check:')
print(df[['Gender', 'Age', 'City', 'CGPA', 'Depression']].head())
```
