## Filtering Data from given DataFrame
    - working on insights.

```python
import pandas as pd

df = pd.read_csv("student_depression_dataset.csv")
```

```python
sample_df = df.sample(5)
```

```python
# Filter Gender male and age <24 and city is pune
df[(df['Gender'] == 'Male') & (df['Age'] < 24) & (df['City'] == 'Pune')]
```

- We must have the insight from the dtypes attribute to filter the records.
    - Same-looking data can behave differently:
        | Value   | dtype  | Behavior           |
        | ------- | ------ | ------------------ |
        | `"100"` | string | text comparison    |
        | `100`   | int    | numeric comparison |

- `strip()`
    - Helps to remove any additional spaces or anything you want to remove from your data.

```python
df['Sleep Duration'].unique()
```

```python
df['Sleep Duration'] = df['Sleep Duration'].str.strip("'")
```

```python
df['Sleep Duration'].unique()
```

## Adding New Column

   - Syntax:
        - df['New Column'] = 'Scalar Value'
        - df['New Column'] = df['Column 1'] + df['Column 2']
                -

```python
df['Bio'] = str(df['Age'])+df['Gender']
```

```python
df.sample(5)
```

## Drop column
   - Syntax:

```python
df.drop('Bio', axis=1, inplace=True)
```
