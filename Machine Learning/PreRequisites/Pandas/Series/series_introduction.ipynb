# Pandas Series

A `Series` is a one-dimensional labeled array in pandas. It is similar to a Python list, but each value has an index label attached to it.

## Why Series is important

- It is one of the core pandas data structures.
- It is useful for working with a single column of data.
- It supports fast operations such as filtering, math, and aggregation.
- It is the building block for many pandas workflows.


## Key points about Series

- A Series has two main parts:
  - `values`: the actual data stored in the Series.
  - `index`: the labels used to access each value.
- All values in a Series usually have the same data type.
- You can create a Series from:
  - a Python list
  - a NumPy array
  - a dictionary
- A Series can also have custom index labels.


## Creating a Series

Below are a few common ways to create a Series in pandas.


```python
import pandas as pd
```

```python
# Series from a list
marks = pd.Series([85, 90, 78, 92])
print('Series from list:')
print(marks)
```

```python
# Series with custom index
students = pd.Series([85, 90, 78, 92], index=['Asha', 'Ravi', 'Neha', 'Karan'])
print('\nSeries with custom index:')
print(students)
```

```python
# Series from a dictionary
scores = pd.Series({'Asha': 85, 'Ravi': 90, 'Neha': 78, 'Karan': 92})
print('\nSeries from dictionary:')
print(scores)
```

```python
# Series with data, custom index, and custom name
ratings = pd.Series(data=[4.5, 4.7, 4.2, 4.9], index=['Asha', 'Ravi', 'Neha', 'Karan'], name='Ratings')
print('\nSeries with custom name:')
print(ratings)
```

## Specifying `dtype`

You can explicitly set the data type of a Series using the `dtype` parameter.

- `int64` is useful for whole numbers.
- `float64` is useful for decimal values.
- `object` is commonly used for text data.


```python
# Series with dtype=int64
numbers = pd.Series([10, 20, 30, 40], dtype='int64')
print('\nSeries with int64 dtype:')
print(numbers)
print('dtype:', numbers.dtype)
```

```python
# Series with dtype=float64
temperature = pd.Series([36.5, 37.2, 38.0, 36.8], dtype='float64')
print('\nSeries with float64 dtype:')
print(temperature)
print('dtype:', temperature.dtype)
```

## Common Series operations

- Access values using index labels or positions.
- Perform arithmetic operations on the whole Series.
- Filter data using conditions.
- Use built-in methods like `mean()`, `max()`, and `min()`.

### Examples

- `marks[0]` gives the first value.
- `students['Asha']` gives the value for Asha.
- `marks.mean()` calculates the average.
- `marks[marks > 80]` filters values greater than 80.


```python
print('Average marks:', marks.mean())
print('Maximum marks:', marks.max())
print('Marks above 80:')
print(marks[marks > 80])

print('\nAccess by label:')
print("Asha's score:", students['Asha'])

print('\nAdd 5 to every value:')
print(marks + 5)
```

## Create a Series from a NumPy array

A NumPy array can be converted into a pandas Series very easily.


```python
import numpy as np

arr = np.array([100, 200, 300, 400])
series_from_array = pd.Series(arr)

print('Series from NumPy array:')
print(series_from_array)
```

## Create a Series from a dictionary

A dictionary becomes a Series where keys are used as the index and values become the data.


```python
data_dict = {'Asha': 85, 'Ravi': 90, 'Neha': 78, 'Karan': 92}
series_from_dict = pd.Series(data_dict)

print('Series from dictionary:')
print(series_from_dict)
```

## Create a Series from a scalar value

A scalar value is repeated for each index label when you provide an index.


```python
scalar_series = pd.Series(5, index=['a', 'b', 'c', 'd'])

print('Series from scalar value:')
print(scalar_series)
```
