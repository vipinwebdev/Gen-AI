# Mathematical Functions and Indexing in Pandas Series

This notebook is a structured study guide for working with mathematical functions and indexing in a pandas `Series`.

A `Series` is a one-dimensional labeled array. It is commonly used to work with a single column of data, perform calculations, and access values by index.


## Learning Goals

- Understand what a pandas `Series` is.
- Use common mathematical functions on a `Series`.
- Learn indexing techniques.
  - Label-based access
  - Position-based access
  - Slicing
  - Boolean filtering
- Learn how to update values using indexing.
- Understand the formulas behind common summary functions.


## Base Data

We will use a few small Series objects throughout the notebook so the examples stay consistent and easy to follow.


```python
import pandas as pd

marks = pd.Series([85, 90, 78, 92, 88], index=['Asha', 'Ravi', 'Neha', 'Karan', 'Priya'])
print(marks)
```

## Mathematical Functions

These functions summarize or describe the values stored in a Series.

- `sum()` adds all values.
- `count()` counts non-missing values.
- `mean()` calculates the average.
- `median()` returns the middle value after sorting.
- `mode()` returns the most frequent value.
- `min()` and `max()` return the smallest and largest values.
- `idxmin()` and `idxmax()` return the index labels of the smallest and largest values.
- `var()` and `std()` measure spread.
- `describe()` gives a compact statistical summary.


## Function Formulas

For the sample Series below:

`[10, 20, 20, 30, 40, 50, 50]`

- `sum()` = `x1 + x2 + ... + xn`
- `count()` = number of valid values
- `mean()` = `sum of values / number of values`
- `median()` = middle value after sorting
  - If the number of values is odd, the middle value is the median.
  - If the number of values is even, the median is the average of the two middle values.
- `mode()` = most frequent value
- `std()` = `sqrt(var)` `->` `standard_deviation = |mean-val1|+|mean - val2|+..../total count of values`
- `var()` = `sum((x - mean)^2) / (n - 1)` `->` `standard_deviation ** 2`


```python
stats = pd.Series([10, 20, 20, 30, 40, 50, 50], index=['a', 'b', 'c', 'd', 'e', 'f', 'g'])
print(stats)
```

```python
print('Sum:', stats.sum())
print('Count:', stats.count())
```

```python
print('Mean:', stats.mean())
```

```python
print('Median:', stats.median())
```

```python
print('Mode:')
print(stats.mode())
```

```python
print('Min value:', stats.min())
print('Max value:', stats.max())
```

```python
print('Index of minimum value:', stats.idxmin())
print('Index of maximum value:', stats.idxmax())
```

```python
print('Variance:', stats.var())
print('Standard deviation:', stats.std())
```

```python
print(stats.describe())
```

## Indexing in a Series

Indexing is the process of selecting data from a Series.

- Label indexing uses the index names.
- Position indexing uses the numeric location.
- Slicing extracts a range of values.
- Boolean indexing filters data based on a condition.


```python
print("Marks for Asha:", marks['Asha'])
```

```python
print('First value:', marks.iloc[0])
```

```python
print('Sliced Series:')
print(marks[1:4])
```

```python
print('Marks greater than 85:')
print(marks[marks > 85])
```

## Updating Values

You can update a Series value using its label.

- This is useful when correcting a value.
- The change happens directly in the Series.


```python
marks['Neha'] = 80
print('Updated Series:')
print(marks)
```

## Key Takeaways

## `describe()` in Detail

`describe()` gives a compact statistical summary of a Series.

- It is very useful when you want to understand data quickly.
- It returns several statistics in one output.
- For a numeric Series, the usual values are:
  - `count`
  - `mean`
  - `std`
  - `min`
  - `25%`
  - `50%` which is the median
  - `75%`
  - `max`
- For non-numeric Series, it may show:
  - `count`
  - `unique`
  - `top`
  - `freq`

### How to read it

- `count`: number of valid values.
- `mean`: average of the values.
- `std`: spread of the values around the mean.
- `min` and `max`: smallest and largest values.
- `25%`, `50%`, `75%`: quartiles that divide the sorted data into parts.


```python
print(stats.describe())

print('\nMeaning of the output for the sample data:')
print('- count: total number of valid values in the Series')
print('- mean: average value of the Series')
print('- std: spread of values around the mean')
print('- min: smallest value')
print('- 25%: first quartile')
print('- 50%: median')
print('- 75%: third quartile')
print('- max: largest value')
```
