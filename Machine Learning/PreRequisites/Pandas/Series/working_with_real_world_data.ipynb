# Working with Real World Data

This notebook explains common pandas `Series` attributes and methods using two real datasets:

- `Global-YouTube-Statistics.csv`
- `Global-YouTube-Statistics_Two_column.csv`


## Load the datasets

We will use the subscriber-related columns from both files to demonstrate the Series features.


```python
import pandas as pd

youtube_full = pd.read_csv('../Global-YouTube-Statistics.csv')
youtube_two = pd.read_csv('../Global-YouTube-Statistics_Two_column.csv')

print(youtube_full.shape)
print(youtube_two.shape)
```

## Series

A pandas `Series` is a one-dimensional labeled array.

- It stores a single column of data.
- It has an index for accessing values.
- It supports vectorized operations and summary functions.


```python
full_subscribers = youtube_full['subscribers']
two_subscribers = youtube_two['subscribers']

print(full_subscribers.head())
print(two_subscribers.head())
```

## `values`

`values` returns the underlying data stored in a Series.

- It is useful when you want the raw array-like representation.
- It does not show the index labels.
- It is often used for quick numeric inspection or conversion.


```python
print(full_subscribers.head().values)
print(two_subscribers.values)
```

## `index`

`index` stores the labels used to access values in a Series.

- For a Series created from a DataFrame column, the index comes from the DataFrame row positions.
- Index labels help you reference values precisely.
- If the index is changed, the access pattern of the Series changes too.


```python
print(full_subscribers.index[:5])
print(two_subscribers.index)
```

## `shape`

`shape` tells you how many elements are in the Series.

- A Series always has one dimension.
- The value returned is a tuple with one number.
- It is useful for checking size before analysis.


```python
print(full_subscribers.shape)
print(two_subscribers.shape)
```

## `ndim`

`ndim` tells you the number of dimensions in the Series.

- A pandas Series is one-dimensional.
- This helps distinguish a Series from a DataFrame.


```python
print(full_subscribers.ndim)
print(two_subscribers.ndim)
```

## `name`

`name` stores the name of the Series.

- It is useful when the Series represents a single logical column.
- The name appears in printed output and summaries.


```python
print(full_subscribers.name)
print(two_subscribers.name)
```

## `is_unique`

`is_unique` checks whether all values in a Series are unique.

- It returns `True` if no value is repeated.
- It returns `False` if any value appears more than once.
- This is useful for spotting repeated categories or IDs.


```python
print(youtube_full['Youtuber'].is_unique)
print(youtube_full['category'].is_unique)
print(youtube_two['Youtuber'].is_unique)
```

## `squeeze()`

`squeeze()` reduces a one-column or one-row object into the simplest possible form.

- If the object has one column, it can become a Series.
- If the object has one value, it can become a scalar.
- It is helpful when a DataFrame is too wide for a Series operation but only one column is needed.


```python
single_col_full = youtube_full[['subscribers']].head()
single_col_two = youtube_two[['subscribers']]

print(single_col_full.squeeze())
print(single_col_two.squeeze())
```

## `head()`

`head()` returns the first few rows or values from a Series.

- The default is the first 5 rows.
- It is commonly used for quick inspection.
- You can pass a number to control how many rows you want.


```python
print(full_subscribers.head(3))
print(two_subscribers.head(3))
```

## `tail()`

`tail()` returns the last few rows or values from a Series.

- The default is the last 5 rows.
- It is useful for checking the end of a dataset.


```python
print(full_subscribers.tail(3))
print(two_subscribers.tail(3))
```

## `sample()`

`sample()` returns random values from a Series.

- It is useful when you want a quick random look at the data.
- The output changes each time unless you set `random_state`.


```python
print(full_subscribers.sample(3, random_state=42))
print(two_subscribers.sample(2, random_state=42))
```

## `value_counts()`

`value_counts()` counts how often each value appears.

- It is very useful for categorical data.
- It helps you see which values are most common.
- It is often used for categories like country, channel type, or labels.


```python
print(youtube_full['category'].value_counts().head())
print(youtube_two['subscribers'].value_counts())
```

## `sort_values()`

`sort_values()` sorts the values in a Series.

- By default, it sorts in ascending order.
- You can change the order using `ascending=False`.
- It is useful for ranking and comparison.


```python
print(full_subscribers.sort_values(ascending=False).head())
print(two_subscribers.sort_values(ascending=False))
```

## `copy()`

`copy()` creates an independent copy of a Series.

- Changes made to the copy do not affect the original.
- This is useful when you want to modify data safely.


```python
full_copy = full_subscribers.head(3).copy()
two_copy = two_subscribers.copy()

full_copy.iloc[0] = -1
two_copy.iloc[0] = -1

print(full_copy)
print(two_copy)
print(full_subscribers.head(3))
print(two_subscribers.head())
```

## Method Chaining

Method chaining means calling multiple Series methods in one expression.

- It keeps code compact.
- It is useful for readable data transformation steps.
- You should still keep it understandable for learning and debugging.


```python
print(youtube_full['subscribers'].dropna().sort_values(ascending=False).head(5))
print(youtube_two['subscribers'].copy().sort_values(ascending=False).head(5))
```

## Key takeaways

- `values`, `index`, `shape`, `ndim`, `name`, and `is_unique` are useful Series attributes.
- `squeeze()`, `head()`, `tail()`, `sample()`, `value_counts()`, `sort_values()`, and `copy()` are practical Series methods.
- Real datasets are a good way to learn these features because they show how Series work in practice.
