# Categorical Plots

- Categorical plots are used to visualize and compare data across different categories.They help in understanding how a numerical variable changes with respect to a categorical variable.

- These plots are mainly used for bivariate analysis, where one variable is categorical and the other is numerical.

- Common types of categorical plots:
    - Bar Plot
    - Count Plot
    - Box Plot
    - Violin Plot

### Bar Plot
- Used to show the average (or aggregated) value of a numerical variable for each category.
- Syntax ---> sns.barplot()
- Parameters:
    - data ---> DataFrame
    - x ---> categorical column
    - y ---> numerical column
    - hue ---> additional categorical grouping
    - estimator ---> function used for aggregation (default = mean)
    - ci ---> shows confidence interval

### Count Plot
- Used to count the number of observations in each category.
- Syntax ---> sns.countplot()
- Parameters:
    - data ---> DataFrame
    - x ---> categorical column
    - hue ---> additional categorical grouping
    - order ---> controls the order of categories

### Box Plot
- Used to show the distribution of data along with median, quartiles, and outliers.
- Syntax ---> sns.boxplot()
- Parameters:
    - data ---> DataFrame
    - x ---> categorical column
    - y ---> numerical column
    - hue ---> additional grouping
    - showfliers ---> show or hide outliers (True/False)

### Violin Plot
- Combines box plot and KDE to show distribution and density.
- Syntax ---> sns.violinplot()
- Parameters:
    - data ---> DataFrame
    - x ---> categorical column
    - y ---> numerical column
    - hue ---> additional grouping
    - split ---> splits the violin for comparison (True/False)

Categorical plots are useful for comparing different categories and understanding patterns in data.
They help in analyzing distributions, counts, and relationships between categorical and numerical variables.
