#   Seaborn

---
Seaborn is a high-level Python data visualization library built on top of Matplotlib. It provides a more attractive and convenient way to create statistical graphics with less code and better default styling.

- Less line of code is required as compare to Matplotlib.
- By default, Seaborn plots looks classic and that also without even bringing any additional customization.

# Purpose of Seaborn

- Seaborn is designed specifically for statistical data visualization. While Matplotlib focuses on flexibility, Seaborn focuses on:
   -   Simplicity
   -   Better visual aesthetics
   -   Built-in statistical insights


# Key Features
   - High-Level Interface
        -   Seaborn reduces the complexity of plotting.
        -   Less code compared to Matplotlib
        -   Automatic handling of labels, legends, and colors
   - Built-in Themes & Styling
        -   Default plots are visually appealing
        -   Supports themes like darkgrid, whitegrid, dark, etc.
        -   No need to manually style plots
   - Works Well with Pandas
        -   Directly works with Pandas DataFrames
        -   Uses column names instead of manual array handling
   - Statistical Visualization
        - Seaborn is especially useful for:
            - Distribution analysis
            - Relationship analysis
            - Categorical comparisons

       - It automatically includes:
            - Confidence intervals
            - Regression lines
            - Aggregations (mean, median)

### 1. Distribution Plots
Used to understand how data is spread.
Examples:
- Histogram
- KDE Plot
- Box Plot

### 2. Relational Plots
Used to find relationships between variables.
Examples:
- Scatter Plot
- Line Plot

### 3. Categorical Plots
Used to compare different categories.
Examples:
- Bar Plot
- Count Plot
- Violin Plot

### 4. Matrix Plots
Used for matrix-like data visualization.
Example:
- Heatmap (used for correlation)

## Seaborn vs Matplotlib

| Feature | Seaborn | Matplotlib |
|--------|--------|------------|
| Ease of use | Easy | Moderate |
| Code length | Short | Longer |
| Styling | Attractive by default | Manual styling |
| Statistical plots | Yes | Limited |

## When to Use Seaborn?

- When you want quick and clean visualizations
- When working with Pandas DataFrames
- When analyzing data distributions and relationships

## Conclusion

Seaborn makes data visualization simple, clean, and visually appealing.
It is especially useful for statistical analysis and is widely used in data science projects.

# plotly
---
- Plotly is a powerful Python library used for creating interactive, web-based graphs and dashboards. Unlike static libraries, Plotly charts are dynamic—you can zoom, hover, filter, and explore data directly in the graph.
- It is widely used in data science, analytics, and dashboards.

# What makes Plotly special?
- Interactive charts (zoom, hover, pan)
- Works in Jupyter Notebook, web apps, and dashboards
- Supports complex visualizations (3D, maps, animations)
- Easy integration with dashboards using Dash

# When to Use Plotly

- Use Plotly when:
    - You want interactive graphs
    - You are building dashboards
    - You need user-friendly visualizations
    - You want to share graphs via HTML

# Key Features
- Interactivity
    - Hover to see exact values
    - Zoom in/out
    - Toggle data series


- Wide Range of Charts
    - Plotly supports:
        - Line charts
        - Scatter plots
        - Bar charts
        - Pie charts
        - Heatmaps
        - 3D plots

- Works with DataFrames
    - Directly works with Pandas
    - No need for manual data conversion

- Browser-Based Output
    - Graphs open in browser or notebook
    - Can be exported as HTML (very useful for sharing)

```python
# plotly example
import plotly.express as px
import pandas as pd
import plotly.io as pio

# Fix renderer
pio.renderers.default = 'browser'
data = {
    "day": [1, 2, 3, 4, 5],
    "sales": [100, 150, 130, 170, 200]
}

df = pd.DataFrame(data)

fig = px.line(df, x="day", y="sales", title="Sales Trend")
fig.show()
```
