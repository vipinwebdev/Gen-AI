
# Regression Plots and Multi-plots

- Regression plots are used to visualize the relationship between variables along with a fitted regression line. They help in understanding trends, correlations, and predictions.
- It shows the relationship between two numerical variables (x and y) along with thw best fit regression line.
- It is a combination of line and scatter plots.
---
- Multi-plots are used to display multiple plots together for better comparison and analysis.

# Regression Plot
 - Used to show the relationship between two variables along with a regression line.

- Syntax ---> sns.regplot()
- Parameters:
    - data ---> DataFrame
    - x ---> independent variable
    - y ---> dependent variable
    - scatter ---> shows or hides scatter points (True/False)
    - fit_reg ---> shows regression line (True/False)
    - ci ---> confidence interval
    - color ---> color of plot



# Multi Plots

### Joint Plot
- Combines scatter plot with distribution plots (histogram/KDE).
- Syntax ---> sns.jointplot()
- Parameters:
    - data ---> DataFrame
    - x ---> variable for x-axis
    - y ---> variable for y-axis
    - kind ---> type of plot (scatter, kde, hex, reg)

- Distribution of multiple numerical variable → Histogram
- Relationship between two numerical column as well → Scatter plot.




### Pair Plot
- Used to plot relationships between all numerical variables in a dataset.
- Syntax ---> sns.pairplot()
- Parameters:
    - data ---> DataFrame
    - hue ---> grouping variable
    - diag_kind ---> type of plot on diagonal (hist/kde)
