#   Bar Chart
---
- Bar plot is being used for categorical - numerical columns.
    -   **Categorical data (labels)** → like names, cities, months, products.
    -   **Numerical values** → like sales, count, marks, etc.
-   In Matplotlib, a bar chart helps you compare values across different categories.
---

-   Simple understanding:
    -   X-axis → categories (e.g., Months: Jan, Feb, Mar)
    -   Y-axis → numerical values (e.g., Sales: 100, 200, 150)

---
-   Why we use bar charts:
    -   To easily compare differences between categories
    -   To visualize distribution of values clearly
    -   Useful when data is not continuous, but grouped

---
-   In Matplotlib:
    -   We use:
        - plt.bar() → for vertical bar chart
        - plt.barh() → for horizontal bar chart


# Stacked Bar Chart
---
-   instead of keeping bars in side by side direction, we can keep the bars in the stacked manner i.e, one on top of other
    in order to figure out which category is having maximum sale, we can have a look over the length of bar.
-   Stacked bar chart ---> bar() ---> parameter → bottom → This bottom parameter specifies that what bar I want to keep below my current bar.
- As we want to difference on the basis of bar length value so that's why we have to keep y-value of previous bar at the bottom parameter.

# Histogram

---
- Instead of comparing categories like bar charts, a histogram is used for continuous numerical data.
- It helps us understand the distribution of data (how values are spread).
- The data is divided into bins (ranges), and each bin shows how many values fall into that range.

- Key Concept
    - X-axis → value ranges (bins)
    - Y-axis → frequency (count of values)
---
- Why we use Histogram
    - To see data distribution
    - To identify:
        - Skewness (left/right)
        - Spread of data
    - Peaks (most frequent range)
        - Useful for:
            - Age distribution
            - Salary distribution
            - Marks analysis
---
- Matplotlib Usage
    - Function: plt.hist()
    - Important parameter:
        - bins → number of intervals (default auto, but better to specify)
