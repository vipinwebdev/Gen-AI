## Pie Chart

- Pie chart is being used for `Bivariate` analysis and is being used for relating `Categorical-Numerical` columns but in pie chart we look in to the `Percentage Distribution`.

---
- A pie chart shows how one whole is split into parts.
- Each slice represents a category.

---

- Use it when you want a quick share or percentage view.
- Keep the chart simple. Too many slices make it hard to read.

---

- `labels` gives names to the slices.
- `sizes` holds the values.
- `explode` can pull out one slice for emphasis.

---

- `autopct` is useful when you want percentages shown on the chart.

---
- A pie chart is created using `plt.pie()`.
- It represents proportions of categories as slices of a circle.


```python

from matplotlib import pyplot as plt

sizes = [40, 30, 20, 10]
labels = ['A', 'B', 'C', 'D']

plt.title('Simple Pie Chart')
plt.pie(sizes, labels=labels)
plt.show()
```
