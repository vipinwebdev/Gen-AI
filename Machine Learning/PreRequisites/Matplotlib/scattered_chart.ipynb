## Scattered Plot/Chart

- A scatter plot is a type of graph used to show the relationship between two numerical variables. It uses dots to represent data points, where each dot shows one observation.
- One variable is placed on the x-axis and the other on the y-axis.

- Why we use scatter plot:
    - Scatter plots help us understand how two variables are related.
    - We can easily see if one variable increases or decreases with the other.
- When to use:
    - We use a scatter plot when we have two numerical values and we want to check their relationship.
---
- For example:
    - Study hours vs marks
    - Temperature vs ice cream sales
---
- What we can learn from it:
    - Positive relationship: both values increase together
    - Negative relationship: one increases while the other decreases
    - No relationship: points are randomly scattered
    - Outliers: values that are very different from others
---
- In simple terms, scatter plot helps us quickly see patterns, trends, and unusual values in the data.
