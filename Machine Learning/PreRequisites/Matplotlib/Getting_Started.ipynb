## What is Data Visualization?
- Data visualization in Python is the process of converting data into graphical representations using libraries like Pandas and Matplotlib to analyze and understand patterns, trends, and insights effectively.

## Common Types of Visualizations
-   | Graph      | Use Case           |
    | ---------- | ------------------ |
    | Bar Chart  | Compare categories |
    | Line Chart | Trends over time   |
    | Histogram  | Data distribution  |
    | Pie Chart  | Proportions        |

    - Example:
        - `import matplotlib.pyplot as plt`
        - `plt.plot(data)`

    - Analysis of one variable at a time to understand its distribution and characteristics.
- Bivariate Analysis
    - Analysis of two variables together to understand the relationship between them.
- Multivariate Analysis
    - Analysis of more than two variables at the same time to understand complex relationships and patterns.

```python
# 2D line PLOT
y = df["Gender"].value_counts()
x = y.index
plt.plot(x, y)
```

- Graph Customization

```python
# Customization
plt.title("Gender Distribution",color="red",fontsize=16)
plt.xlabel("Gender",color="green")
plt.ylabel("Number of Students",color='yellow')
plt.grid()
plt.plot(x, y)
plt.figure(figsize=(10, 6))
```

- line Customization
    - Markers => `o, ^, s, +, *, d, h, p`
        | Marker | Shape         | Meaning                    |
        | ------ | ------------- | -------------------------- |
        | `o`    | ● Circle      | Most commonly used         |
        | `^`    | ▲ Triangle Up | Points upward              |
        | `s`    | ■ Square      | Box-shaped marker          |
        | `+`    | + Plus        | Cross lines                |
        | `*`    | ★ Star        | Highlight important points |
        | `d`    | ◆ Diamond     | Tilted square              |
        | `h`    | ⬢ Hexagon     | 6-sided shape              |
        | `p`    | ⬟ Pentagon    | 5-sided shape              |


```python
plt.figure(figsize=(10, 6))
plt.title("Gender Distribution",color="red",fontsize=16)
plt.xlabel("Gender",color="green")
plt.ylabel("Number of Students",color='green')

plt.grid()
# Default marker-size = 6
plt.plot(x, y, color="red",linewidth=2, marker='o', linestyle='-.', markersize=10) # line-style => '--', '-.'
plt.show()
```

## What is groupby() in Pandas?

    - groupby() is used to split data into groups, apply some operation (like sum, count, avg), and then combine the results.

        - # groupby():
            # Used to group data based on a column (e.g., Country)
            # Then apply aggregation (sum, mean, etc.)

            # In this example:
            # India → 150 + 5000 = 5150
            # China → 5000
            # USA → 1000 + 4000 = 5000

            # Syntax:
            # df.groupby('column_name')['value_column'].aggregation_function()

- To convert a Date column to datetime you can use pd.to_datetime()
- To extract the Year from the DateTime column ---> df['col_name'].dt.year
- To extract the Month value (1-12) from the DateTime column ---> df['col_name'].dt.month
- To extract the Month name from the DateTime column ---> df['col_name'].dt.month_name()

```python
import pandas as pd

data = {
    'Date': ['2024-01-15', '2024-03-20', '2024-07-10']
}

df = pd.DataFrame(data)

# Convert to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract components
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Month_Name'] = df['Date'].dt.month_name()

df
```
