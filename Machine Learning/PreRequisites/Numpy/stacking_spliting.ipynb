# NumPy Stacking and Splitting

This notebook contains complete notes on NumPy stacking and splitting.

It covers:
- the main types of stacking and splitting
- what each function does
- when to use each function
- separate examples for 1D, 2D, and 3D arrays


## 1. Stacking

Stacking means combining arrays to form a larger array.

Common stacking functions:
- `np.concatenate()`
- `np.vstack()`
- `np.hstack()`
- `np.dstack()`
- `np.stack()`
- `np.column_stack()`


### Types of stacking

| Function | Meaning | Typical use |
|---|---|---|
| `concatenate` | Joins arrays along an existing axis | General purpose joining |
| `vstack` | Stacks arrays vertically | Row-wise stacking |
| `hstack` | Stacks arrays horizontally | Column-wise stacking |
| `dstack` | Stacks arrays along depth axis | 3D style stacking |
| `stack` | Joins arrays along a new axis | Creates a new dimension |
| `column_stack` | Stacks 1D arrays as columns | Building 2D tabular data |


### `np.concatenate()`

- Joins two or more arrays along an existing axis.
- Does not create a new axis, so the input shapes must match on all other axes.
- Is the most general stacking/joining function in NumPy.
- Commonly used when combining datasets, rows, or columns that already share compatible shapes.
- Example use: merging two 1D arrays into one longer array or joining 2D arrays row-wise or column-wise.

Syntax:
```python
np.concatenate((array1, array2), axis=0)
```


````python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = np.concatenate((a, b))
print(result)
````

### `np.vstack()`

- Stacks arrays vertically, one array on top of another.
- Treats 1D inputs as row vectors before stacking.
- Is useful when you want to add more samples/rows to a matrix.
- Requires the arrays to have matching column counts for 2D inputs.
- Internally it is a convenience wrapper around `concatenate` with row-wise behavior.


````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = np.vstack((a, b))
print(result)
````

### `np.hstack()`

- Stacks arrays horizontally, side by side.
- For 1D arrays, it simply joins values end to end.
- For 2D arrays, it adds columns to the right of the existing array.
- Is useful when combining feature columns or appending values across the width of an array.
- All arrays must be compatible along the non-stacking dimensions.


````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = np.hstack((a, b))
print(result)
````

### `np.dstack()`

- Stacks arrays along the third axis, also called the depth axis.
- Converts lower-dimensional inputs into a 3D-style structure when possible.
- Is useful when you want to keep different arrays as separate layers.
- Often used in image or multi-channel style data manipulation.
- All arrays must match on the dimensions that are not being stacked.


````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = np.dstack((a, b))
print(result)
````

### `np.stack()`

- Joins arrays by creating a new axis.
- Is different from `concatenate` because it increases the number of dimensions.
- Lets you choose where the new axis is inserted with the `axis` parameter.
- Is useful when you want to preserve arrays as separate samples or layers.
- All input arrays must have exactly the same shape.


````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = np.stack((a, b), axis=0)
print(result)
````

### `np.column_stack()`

- Stacks 1D arrays as columns to form a 2D matrix.
- Is similar to horizontal stacking, but is especially convenient for 1D inputs.
- Automatically converts 1D arrays into column-ready form.
- Is commonly used when building table-like data from separate vectors.
- All arrays should have the same length.


````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = np.column_stack((a, b))
print(result)
````

## 1D Array Stacking Examples

These examples use one-dimensional arrays.


````python
a = np.array([10, 20, 30])
b = np.array([40, 50, 60])

print('concatenate:', np.concatenate((a, b)))
print('vstack:\n', np.vstack((a, b)))
print('hstack:', np.hstack((a, b)))
print('stack axis=0:\n', np.stack((a, b), axis=0))
print('column_stack:\n', np.column_stack((a, b)))
````

## 2D Array Stacking Examples

These examples use two-dimensional arrays.


````python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

print('concatenate axis=0:\n', np.concatenate((a, b), axis=0))
print('concatenate axis=1:\n', np.concatenate((a, b), axis=1))
print('vstack:\n', np.vstack((a, b)))
print('hstack:\n', np.hstack((a, b)))
print('stack axis=0:\n', np.stack((a, b), axis=0))
````

## 3D Array Stacking Examples

These examples use three-dimensional arrays.


````python
a = np.array([[[1, 2], [3, 4]]])
b = np.array([[[5, 6], [7, 8]]])

print('concatenate axis=0:\n', np.concatenate((a, b), axis=0))
print('concatenate axis=1:\n', np.concatenate((a, b), axis=1))
print('concatenate axis=2:\n', np.concatenate((a, b), axis=2))
print('stack axis=0 shape:', np.stack((a, b), axis=0).shape)
print('dstack shape:', np.dstack((a, b)).shape)
````

## 2. Splitting

Splitting means breaking one array into multiple smaller arrays.

Common splitting functions:
- `np.split()`
- `np.array_split()`
- `np.vsplit()`
- `np.hsplit()`
- `np.dsplit()`


### Types of splitting

| Function | Meaning | Typical use |
|---|---|---|
| `split` | Splits array into equal parts | Exact divisions only |
| `array_split` | Splits array into uneven parts if needed | Flexible splitting |
| `vsplit` | Splits vertically | Row-wise split |
| `hsplit` | Splits horizontally | Column-wise split |
| `dsplit` | Splits along depth axis | 3D split |


### `np.split()`

- Splits an array into equal parts.
- Requires the split to divide the array evenly unless the axis length supports the requested sections.
- Returns a list of sub-arrays.
- Is useful when you need fixed-size chunks for batch processing or testing.
- Works on any axis, not only on 1D arrays.


````python
a = np.array([1, 2, 3, 4, 5, 6])

result = np.split(a, 3)
print(result)
````

### `np.array_split()`

- Splits an array into parts even when the size is not perfectly divisible.
- Is more flexible than `np.split()` because it can create uneven sections.
- Returns a list of sub-arrays.
- Is useful when you want approximate chunk sizes without raising an error.
- Works well for slicing datasets into train, validation, or inspection groups.


````python
a = np.array([1, 2, 3, 4, 5])

result = np.array_split(a, 3)
print(result)
````

### `np.vsplit()`

- Splits an array vertically along rows.
- Is mainly used for 2D and higher-dimensional arrays.
- Requires the row count to be divisible by the number of splits when using equal splitting.
- Is useful when you want to break a matrix into upper and lower parts.
- Produces a list of smaller arrays.


````python
a = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])

result = np.vsplit(a, 2)
print(result)
````

### `np.hsplit()`

- Splits an array horizontally along columns.
- Is mainly used for 2D and higher-dimensional arrays.
- Requires the column count to be compatible with the requested split.
- Is useful when separating features or extracting left/right portions of a matrix.
- Produces a list of smaller arrays.


````python
a = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])

result = np.hsplit(a, 2)
print(result)
````

### `np.dsplit()`

- Splits an array along the depth axis.
- Is intended for 3D arrays or arrays that can be viewed as 3D-like structures.
- Separates data into depth slices or layers.
- Is useful for channel-based or volumetric data.
- Produces a list of smaller arrays.


````python
a = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

result = np.dsplit(a, 2)
print(result)
````

## 1D Array Splitting Examples

These examples use one-dimensional arrays.


````python
a = np.array([1, 2, 3, 4, 5, 6])

print('split:', np.split(a, 3))
print('array_split:', np.array_split(a, 4))
````

## 2D Array Splitting Examples

These examples use two-dimensional arrays.


````python
a = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])

print('vsplit:', np.vsplit(a, 2))
print('hsplit:', np.hsplit(a, 2))
````

## 3D Array Splitting Examples

These examples use three-dimensional arrays.


````python
a = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

print('dsplit:', np.dsplit(a, 2))
print('split axis=0:', np.split(a, 2, axis=0))
````

## Quick Summary

- Use stacking to combine arrays.
- Use splitting to divide arrays.
- Use `concatenate` when joining along an existing axis.
- Use `stack` when you want a new axis.
- Use `array_split` when an equal split is not possible.
