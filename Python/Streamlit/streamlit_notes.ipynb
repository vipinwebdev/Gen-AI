# Streamlit Notes

Streamlit is a Python framework used to build interactive web applications for data science, machine learning, dashboards, and quick internal tools.

Its main advantage is simplicity: you write normal Python code, and Streamlit automatically turns it into a web app.


## 1. What Is Streamlit?

Streamlit is designed for developers who want to create web interfaces without dealing with HTML, CSS, JavaScript, or backend frameworks.

Common uses:
- data visualization dashboards
- machine learning demo apps
- reporting tools
- internal business tools
- quick prototypes for user interfaces

A Streamlit app runs as a Python script. Each time the user interacts with the app, the script is re-executed from top to bottom.


## 2. Why Use Streamlit?

Key benefits:
- very fast to build and test
- pure Python syntax
- built-in widgets and charts
- easy integration with pandas, NumPy, matplotlib, Plotly, and machine learning libraries
- useful for both beginners and professionals

Limitations:
- less flexible than full frontend frameworks for highly customized UI
- app execution model can confuse beginners because the whole script reruns after interaction
- large apps need careful design for state and performance


## 3. Installation and Running a Streamlit App

Install Streamlit using pip:

```bash
pip install streamlit
```

Run an app file:

```bash
streamlit run app.py
```

After running the command, Streamlit starts a local server and opens the app in the browser.


## 4. Basic Structure of a Streamlit App

A simple Streamlit app looks like this:


````python
import streamlit as st

st.title("My First Streamlit App")
st.header("Welcome")
st.write("This is a simple Streamlit application.")
````

### Explanation

- `st.title()` creates the main page title.
- `st.header()` creates a section heading.
- `st.write()` is a flexible function that can display text, variables, tables, and more.

`st.write()` is one of the most commonly used functions because it automatically handles different data types.


## 5. Displaying Text and Content

Streamlit provides several functions to display text in different formats.


````python
import streamlit as st

st.title("Streamlit Text Examples")
st.header("Header Example")
st.subheader("Subheader Example")
st.text("This is plain text.")
st.markdown("**This is bold text using Markdown.**")
st.write("This text is displayed using st.write().")
st.caption("This is a caption.")
st.code("print('Hello Streamlit')", language="python")
st.latex(r"a^2 + b^2 = c^2")
````

## 6. Streamlit Widgets

Widgets allow users to interact with the app. These widgets return values that you can use in your Python logic.

Common widgets:
- `st.button()`
- `st.checkbox()`
- `st.radio()`
- `st.selectbox()`
- `st.multiselect()`
- `st.slider()`
- `st.text_input()`
- `st.text_area()`
- `st.number_input()`
- `st.date_input()`
- `st.file_uploader()`


````python
import streamlit as st

name = st.text_input("Enter your name")
age = st.number_input("Enter your age", min_value=1, max_value=100, value=25)
gender = st.radio("Select gender", ["Male", "Female", "Other"])
skills = st.multiselect("Select skills", ["Python", "SQL", "Excel", "Machine Learning"])
experience = st.slider("Years of experience", 0, 20, 2)

st.write("Name:", name)
st.write("Age:", age)
st.write("Gender:", gender)
st.write("Skills:", skills)
st.write("Experience:", experience)
````

### Important Point About Buttons

`st.button()` returns `True` only on the run in which the button is clicked.

Example:


````python
import streamlit as st

if st.button("Click Me"):
    st.success("Button was clicked")
````

## 7. Layout and Page Organization

Streamlit provides layout tools to organize the page better.

Useful layout functions:
- `st.sidebar`
- `st.columns()`
- `st.tabs()`
- `st.expander()`
- `st.container()`


````python
import streamlit as st

st.sidebar.title("Sidebar Menu")
choice = st.sidebar.selectbox("Choose a page", ["Home", "About", "Contact"])

col1, col2 = st.columns(2)
with col1:
    st.write("This is column 1")
with col2:
    st.write("This is column 2")

with st.expander("See more details"):
    st.write("Expanded content goes here.")
````

## 8. Displaying DataFrames, Metrics, and Charts

Streamlit works very well with tabular data and visualizations.


````python
import streamlit as st
import pandas as pd

data = pd.DataFrame(
    {
        "Month": ["Jan", "Feb", "Mar", "Apr"],
        "Sales": [120, 150, 170, 160]
    }
)

st.dataframe(data)
st.table(data)
st.metric(label="Total Sales", value=data["Sales"].sum())
st.line_chart(data.set_index("Month"))
````

### Common Data Display Functions

- `st.dataframe()` shows interactive tables.
- `st.table()` shows a static table.
- `st.metric()` highlights a key business number.
- `st.line_chart()`, `st.bar_chart()`, and `st.area_chart()` create quick charts.
- Streamlit also supports matplotlib, Plotly, Altair, and seaborn visualizations.


## 9. Streamlit Execution Model

This is one of the most important concepts in Streamlit.

Whenever a user changes a widget value, Streamlit reruns the entire script from top to bottom.

This means:
- variables are recreated on every run unless stored properly
- widget values control the flow of the app
- session state is needed to remember values across reruns


## 10. Session State

`st.session_state` helps store values across reruns.

It is useful for:
- counters
- login information
- multi-step forms
- remembering previous choices


````python
import streamlit as st

if "count" not in st.session_state:
    st.session_state.count = 0

if st.button("Increment"):
    st.session_state.count += 1

st.write("Current count:", st.session_state.count)
````

## 11. Forms in Streamlit

Normally, every widget interaction reruns the app immediately. A form groups multiple inputs and submits them together.

This is useful when you want users to enter several values before processing them.


````python
import streamlit as st

with st.form("user_form"):
    username = st.text_input("Username")
    password = st.text_input("Password", type="password")
    submitted = st.form_submit_button("Submit")

if submitted:
    st.success(f"Welcome, {username}")
````

## 12. File Uploading

Streamlit can upload files directly from the UI.

This is especially useful for CSV files, Excel sheets, images, and documents.


````python
import streamlit as st
import pandas as pd

uploaded_file = st.file_uploader("Upload a CSV file", type=["csv"])

if uploaded_file is not None:
    df = pd.read_csv(uploaded_file)
    st.write("Preview of uploaded data:")
    st.dataframe(df.head())
````

## 13. Caching in Streamlit

Because the script reruns often, expensive computations may repeat unnecessarily. Streamlit provides caching to improve performance.

Two common decorators:
- `@st.cache_data` for data loading and transformations
- `@st.cache_resource` for heavy resources like models or database connections


````python
import streamlit as st
import pandas as pd

@st.cache_data
def load_data():
    data = pd.DataFrame(
        {
            "Name": ["A", "B", "C"],
            "Score": [90, 85, 88]
        }
    )
    return data

df = load_data()
st.dataframe(df)
````

## 17. Multipage Apps

Streamlit supports multipage apps by placing separate Python files inside a `pages` folder.

Example structure:

```text
project_folder/
├── app.py
└── pages/
    ├── 1_Home.py
    ├── 2_Analytics.py
    └── 3_Settings.py
```

This makes larger apps easier to organize and maintain.


## 18. Deployment Options

A Streamlit app can be deployed in several ways:
- Streamlit Community Cloud
- local server inside an organization
- Docker container
- cloud platforms such as AWS, Azure, or GCP
- hosting through reverse proxy tools like Nginx

For deployment, projects usually need:
- `requirements.txt`
- the main app file
- any dataset or resource files
- optional `.streamlit/config.toml`


## 19. Best Practices

- keep the app simple and readable
- avoid repeating expensive operations on every rerun
- use `st.session_state` for persistent interaction logic
- use forms when multiple inputs belong together
- separate business logic from UI code when the app grows
- validate uploaded files and user inputs
- use caching for performance improvement
- organize large apps into multiple files or pages


## 20. Mini Streamlit Example

The following example combines several common features in one small app.


````python
import streamlit as st
import pandas as pd

st.set_page_config(page_title="Student Dashboard", layout="wide")

st.title("Student Dashboard")

name = st.text_input("Enter student name")
marks = st.slider("Select marks", 0, 100, 75)

if st.button("Show Result"):
    if marks >= 40:
        st.success(f"{name} has passed with {marks} marks")
    else:
        st.error(f"{name} has failed with {marks} marks")

df = pd.DataFrame(
    {
        "Subject": ["Math", "Science", "English"],
        "Marks": [78, 88, 81]
    }
)

st.subheader("Marks Table")
st.dataframe(df)
st.bar_chart(df.set_index("Subject"))
````
