## Modules and Packages

    # A module is simply a .py file that contains:
        - functions
        - variables
        - classes

        Example: math_utils.py
            def add(a, b):
                return a + b

            def multiply(a, b):
                return a * b

        Using a Module:
            import math_utils
            print(math_utils.add(2, 3))

        # Import Specific Function
            from math_utils import add
            print(add(5, 2))
