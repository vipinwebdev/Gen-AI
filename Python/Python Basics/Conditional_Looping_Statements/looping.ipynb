## Looping Statements
    Loops in Python (for, while) are used to execute a block of code repeatedly over a sequence
    or based on a condition.
        Important Points
        for → Used for iterable data
        while → Used for condition-based loops
        break, continue, pass control loop flow
        range() generates numbers

```python
# for -> Iterates over iterable (list, tuple, string, etc.)
for i in range(5):
    print(i)
```

```python
lst = [10, 20, 30]

# for -> Iterate list elements
for item in lst:
    print(item)
```

```python
name = "Vipin"

# for -> Iterate characters
for ch in name:
    print(ch)
```

```python
i = 0

# while -> Runs until condition becomes False
while i < 5:
    print(i)
    i += 1
```

```python
# Infinite loop -> Runs forever (until break)
while True:
    print("Running...")
    break
```

## Loop Control Statements

```python
# break -> Exit loop when condition met

for i in range(5):
    if i == 3:
        break
    print(i)
```

```python
# continue -> Skip current iteration
for i in range(5):
    if i == 2:
        continue
    print(i)
```

```python
# pass -> No operation
for i in range(5):
    if i == 2:
        pass
    print(i)
```

```python
# range() Function -> Generates sequence of numbers

# range(start, stop, step)
for i in range(1, 10, 2):
    print(i)
```

```python
# Nested loop -> Loop inside loop
for i in range(3):
    for j in range(2):
        print(i, j)
```

```python
# else with Loop
#🔹 Executes when loop completes normally
for i in range(3):
    print(i)
else:
    print("Loop completed")
```

```python
# Loop with enumerate()
#🔹 Get index + value
lst = ["a", "b", "c"]

for i, val in enumerate(lst):
    print(i, val)
```

```python
# zip() -> 🔹 Iterate multiple lists together
names = ["Vipin", "Rahul"]
ages = [25, 26]

for name, age in zip(names, ages):
    print(name, age)
```
