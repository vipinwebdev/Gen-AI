## File Handling
    # File handling in Python allows you to create, read, write, append, and delete files.
    # Python provides built-in functions like:
        ~ open()
        ~ read()
        ~ write()
        ~ close()
      Syntax: Opening a File
        file = open("filename", "mode")

## Modes:
    | Mode  | Meaning             |
    | ----- | ------------------- |
    | `'r'` | Read (default)      |
    | `'w'` | Write (overwrite)   |
    | `'a'` | Append              |
    | `'x'` | Create new file     |
    | `'b'` | Binary mode         |
    | `'t'` | Text mode (default) |


##
    # NLP ---> Text Data (Sequential Data) ---> Text Format File
    # ML Models ---> Cat Dog Image Classifier ---> png of jpeg ---> Binary Format file
     # deep Learning ----> Binary Format Files.

```python
# Reading a File
file = open("data.txt", "r", encoding="utf-8")

content = file.read()
print(content)

file.close()
```

```python
# Other methods:
file = open("data.txt", "r", encoding="utf-8")
print(file.read(10)  )      # Read first 10 characters
file.readline()      # Read one line
file.readlines()     # Read all lines as list
file.close()
```

```python
# Writing to a File

file = open("data.txt", "w",encoding="utf-8")

file.write("Hello Vipin!")
file.close()
```

```python
# Appending to a File
file = open("data.txt", "a")

file.write("\nNew line added")

file.close()
```

```python
# Using 'with' statement for automatic file closing
with open("data.txt", "r", encoding="utf-8") as file:
    content = file.read()
    print(content)
```

```python
# Working with Lines
with open("data.txt", "r") as file:
    for line in file:
        print(line.strip())
```

```python
# Writing Multiple Lines

lines = ["Line 1\n", "Line 2\n", "Line 3\n"]

with open("data.txt", "w") as file:
    file.writelines(lines)
```

```python
# File Pointer (Cursor)

with open("data.txt", "r") as file:
    print(file.tell())   # Current position
    file.read(5)
    print(file.tell())   # Position after reading

    file.seek(0)         # Move to start
```

```python
# File Handling with Binary Files
with open("image.jpg", "rb") as file:
    data = file.read()

```

```python
with open("copy.jpg", "wb") as file:
    file.write(data)
```

```python
# File Handling Example (Real Use Case)

with open("data.txt", "r") as file:
    count = sum(1 for line in file)

print("Total lines:", count)
```

```python
# Advanced: Working with JSON Files

import json

data = {"name": "Vipin", "age": 25}

# Write JSON
with open("data.json", "w") as file:
    json.dump(data, file)

# Read JSON
with open("data.json", "r") as file:
    obj = json.load(file)

print(obj)
```

```python
# CSV File Handling
import csv

with open("data.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Age"])
    writer.writerow(["Vipin", 25])

with open("data.csv", "r", newline="") as file:
    print(file.read())
```

```python
#  Read total lines, words, characters
def count_lines(file):
    with open(file, "r") as f:
        lines = f.readlines()
        total_lines = len(lines)
        total_words = sum(len(line.split()) for line in lines)
        total_chars = sum(len(line) for line in lines)
        return total_lines, total_words, total_chars

total_lines,total_words,total_chars = count_lines("data.txt")
print(total_lines, total_words, total_chars)
```
