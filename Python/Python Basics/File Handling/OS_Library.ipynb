## OS Library
    # The os module in Python provides a way to interact with the Operating System (OS).
    # It allows you to:
        - Work with files & directories
        - Access environment variables
        - Execute system commands
        - Handle paths (cross-platform)

```python
import os

# Get Current Directory
current_path = os.getcwd()
print(current_path)
```

```python
# Change Directory
os.chdir("C:/Users/Vipin/Documents")
```

```python
# List Files in Current Directory
os.listdir()
```

```python
# Create Directory
os.mkdir("test_folder")
```

```python
# Create Nested Directories
os.makedirs("parent/child/grandchild")
```

```python
# Remove Directory
os.rmdir("test_folder")
```

```python
# Remove Nested Directories
os.removedirs("parent/child/grandchild")
```

```python
# Rename File
os.rename("old.txt", "new.txt")
```

```python
# Delete File
os.remove("data.txt")
```

```python
# Join Paths
os.path.join("folder", "file.txt")
```

```python
# Check Path Exists
os.path.exists("data.txt")
```

```python
# Check if File
os.path.isfile("data.txt")
```

```python
# Check if Directory
os.path.isdir("folder")
```

```python
# Get File Size
os.path.getsize("data.txt")
```

```python
# Get File Name
os.path.basename("/home/user/file.txt")
```

```python
# Get Directory Name
os.path.dirname("/home/user/file.txt")
```

```python
# Get Environment Variable
os.environ.get("PATH")
```

```python
# Set Environment Variable
os.environ["MY_VAR"] = "Hello"
```

```python
# Run System Command (Windows)
os.system("dir")
```

```python
# Run System Command (Linux/Mac)
import os
os.system("ls")
```

```python
# Get Process ID
import os
os.getpid()
```

```python
# Get Current User
import os
os.getlogin()
```

```python
# Get OS Type
os.name
```
