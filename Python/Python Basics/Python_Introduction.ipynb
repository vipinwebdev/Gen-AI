## f-String (Formatted String Literal)
    An f-string is a modern way to format strings in Python by embedding expressions inside {}.
        Syntax:
            f"string {expression}"

    f-strings allow embedding variables and expressions directly inside strings using {}.

```python
name = "Vipin"
age = 25

# f-string -> Embed variables directly
result = f"My name is {name} and I am {age} years old"
print(result)
```
