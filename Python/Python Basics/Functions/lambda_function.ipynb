## Lambda function
    - Lambda functions are the one liner functions
    - defined by lambda keyword
    - Syntax:
        lambda arguments : expression
        keyword input : logic

```python
square = lambda x : x*x
print(square(6))
```

```python
## They can have any number of input parameters but only one expression or one logic.
## They are used for short operations or as arguments Higher order functions.
## It can be treated as a function without a name, and they don't even require a return type.

sum_of_two_numbers = lambda x,y : x+y
sum_of_two_numbers(10,20)
```
