## map() Function
    # map() is used to apply a function to each element of an iterable and return a new iterator.
    # Syntax:
        map(function, iterable)


    # map() is particularly useful for transforming the data or applying an opertion on all items in the iterable.

    # map() is going to help us out with avoiding unnecessory lines of code and we can apply the operation on all values or all items in an iterable and this is the sign of a good programmer.


```python
import string


## Basic Example

# map() -> Apply function to each element
def square(x):
    return x * x

nums = [1, 2, 3, 4]
result = map(square, nums)

print(list(result))
```

```python
def find_even_number(num):
    if num % 2 == 0:
        return num
    return None


print(list(map(find_even_number, nums)))
```

```python
def convert_upper_case(string):
        return string.upper()

print(list(map(convert_upper_case, ["string","char"])))
```
