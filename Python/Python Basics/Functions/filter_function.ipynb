## filter()
    # It is used to select or find elements from an iterable based on a condition.
    # filter() selects elements from an iterable based on a condition returning True.

    # Important Points:
        - Returns a filter object (iterator)
        - Convert using list()
        - Used only for filtering (not transformation)
        - Use filter() in functional programming style

    # Real-world Use Cases
        - Filtering API data
        - Removing invalid entries
        - Validating inputs
        - Data processing pipelines

    # Syntax:
        filter(function, iterable)

```python
# filter() -> Select even numbers
def is_even(x):
    return x % 2 == 0

nums = [1, 2, 3, 4, 5, 6]
result = filter(is_even, nums)

print(list(result))
```

```python
nums = [1, 2, 3, 4, 5, 6]

# filter() with lambda
even_numbers = list(filter(lambda x: x % 2 == 0, nums))

print(even_numbers)
```

```python
names = ["Vipin", "Amit", "Raj", "Anu"]

# Select names starting with 'A'
result = list(filter(lambda x: x.startswith("A"), names))

print(result)
```

```python
data = [0, 1, "", "Hello", None, 5]

# Removes falsy values
result = list(filter(None, data))

print(result)
```

## filter() vs List Comprehension

```python
nums = [1, 2, 3, 4, 5, 6]
# Using filter()
list(filter(lambda x: x % 2 == 0, nums))

# Using list comprehension (preferred)
[x for x in nums if x % 2 == 0]
```
