```python
## Create a function that converts temperature from Celsius to Fahrenheit.
## If temperature is in Celsius, then Celsius to Fahrenheit conversion is:
'''
1 deg = 9/5 + 32 F
45 deg = 45 * 9/5 + 32 F

1 F = 5/9 -32 C
45 F = 45 * 5/9 - 32 C
'''

def convert_temperature(temp_value, convert_in):
    if convert_in == 'F':
        print(f"The given temperature is {temp_value} F and after conversion in Celsius is {temp_value*5/9 - 32} C")
    elif convert_in == 'C':
        print(f"The given temperature is {temp_value} C and after conversion in Fahrenheit is {temp_value*9/5 + 32} F")

    else:
        print("The given temperature is invalid")

convert_temperature(1,"C")
convert_temperature(97,"F")
convert_temperature(37,"C")
```

```python
# Define a function that would behave as a Password Strength checker.
'''
Strong password:
    If your password have:
        1. Length is greater than or equal to 8 characters.
        2. Password must consist of at least 1 lower case letters.
        3. Password must consist of at least 1 upper case letters.
        4. Password must consist of at least 1 digit.
        5. Password must consist of at least 1 special characters.
Weak Password:
    If your password doesn't have any of the above points:
'''


def check_password_length(password):
    special_chars= '!@#$^*()_'
    if password == '':
        print("Password cannot be empty.")
    if len(password) < 8:
        print("Password must be at least 8 characters.")
    elif not any (char.islower() for char in password):
        print("Password must consist of at least 1 lower case letters.")
    elif not any(char.upper() for char in password):
        print("Password must consist of at least 1 upper case letters.")
    elif not any(char.isdigit() for char in password):
        print("Password must consist of at least 1 digit.")
    elif not any (char in special_chars for char in password):
        print("Password must consist of at least 1 special characters.")

check_password_length('')
check_password_length('Hello1232')
check_password_length('Hello1232')
check_password_length('Hello1232@')
```

```python
# Define functions to check palindrome

def check_palindrome(string):
    if string == '':
        print("String cannot be empty.")
    else:
        if string == string[::-1]:
            print(f"Given String '{string}' is palindrome.")
        else:
            print(f"Given String '{string}' is not palindrome.")

check_palindrome("naman")
check_palindrome("hello")
check_palindrome("ab")
```
