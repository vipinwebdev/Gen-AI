## Exception Handling

    # Exception handling is a mechanism to handle runtime errors gracefully, preventing program crashes and allowing controlled responses to unexpected situations.

## Why Need Exception Handling?

    # 1. Prevent Program Crashes
    # Without exception handling, programs crash unexpectedly:

    # Without exception handling - CRASHES!
        number = int("not a number")  # ValueError - program crashes!
        print("This never runs")

        # With exception handling - GRACEFUL!
        try:
            number = int("not a number")
        except ValueError:
            print("Invalid input, using default value 0")
            number = 0
        print("Program continues!")  # This runs!

    2. Handle User Input Errors
        def get_user_age():
            while True:
                try:
                    age = int(input("Enter your age: "))
                    if age < 0:
                        raise ValueError("Age cannot be negative")
                    return age
                except ValueError as e:
                    print(f"Invalid input: {e}. Please try again.")

    # User can keep trying without crashing
    age = get_user_age()

    3. Resource Management
        # Ensure resources are properly cleaned up
            file = None
            try:
                file = open("important_data.txt", "r")
                data = file.read()
                # Process data
            except FileNotFoundError:
                print("File not found, creating new file")
                file = open("important_data.txt", "w")
            finally:
                if file:
                    file.close()  # Always close file

    4. Debugging and Logging:

        import logging

        logging.basicConfig(level=logging.ERROR)

        def divide_numbers(a, b):
            try:
                result = a / b
            except ZeroDivisionError as e:
                logging.error(f"Division by zero attempted: {e}")
                return None
            except TypeError as e:
                logging.error(f"Invalid types: {e}")
                return None
            else:
                logging.info(f"Successful division: {a}/{b}={result}")
                return result

## Python Exception Handling Syntax
    # Basic Structure
        try:
            # Code that might raise an exception
            risky_operation()
        except ExceptionType:
            # Handle the specific exception
            handle_error()
        else:
            # Runs if no exception occurred
            print("Success!")
        finally:
            # Always runs (cleanup code)
            cleanup()

```python
def safe_file_reader(filename):
    try:
        file = open(filename, 'r')
        content = file.read()

    except FileNotFoundError:
        print(f"Error: File '{filename}' not found")
        return None

    except PermissionError:
        print(f"Error: Permission denied for '{filename}'")
        return None

    except Exception as e:
        print(f"Unexpected error: {e}")
        return None

    else:
        print(f"Successfully read {filename}")
        return content

    finally:
        try:
            file.close()
            print(f"Closed {filename}")
        except:
            pass  # File might not have been opened

# Usage
result = safe_file_reader("config.txt")
```
