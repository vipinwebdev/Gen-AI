## Advanced Exception Handling

```python
#   1. Multiple Exceptions

try:
    num = int(input("Enter a number: "))
    result = 100 / num
    print(f"Result: {result}")

except (ValueError, ZeroDivisionError) as e:
    print(f"Error: {e}")

# Or separate handlers
except ValueError:
    print("Not a valid number")
except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
#   2. Custom Exceptions
class InsufficientFundsError(Exception):
    """Raised when account balance is insufficient"""
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount = amount
        super().__init__(f"Insufficient funds: ${balance} available, ${amount} requested")

class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def withdraw(self, amount):
        if amount > self.balance:
            raise InsufficientFundsError(self.balance, amount)
        self.balance -= amount
        return self.balance

# Usage
account = BankAccount(100)
try:
    account.withdraw(150)
except InsufficientFundsError as e:
    print(f"Error: {e}")
    print(f"Balance: ${e.balance}, Requested: ${e.amount}")
```

```python
#   3. Exception Chaining
def process_data(data):
    try:
        value = int(data)
        result = 100 / value
    except ValueError as e:
        raise RuntimeError("Failed to convert data to number") from e
    except ZeroDivisionError as e:
        raise RuntimeError("Cannot divide by zero") from e

try:
    process_data("abc")
except RuntimeError as e:
    print(f"Error: {e}")
    print(f"Caused by: {e.__cause__}")
```

```python
#   4. Context Managers (with statement)
# Better resource management
try:
    with open("file.txt", "r") as file:
        content = file.read()
        # Auto-closes file even if exception occurs
except FileNotFoundError:
    print("File not found")

# Create custom context manager
class DatabaseConnection:
    def __enter__(self):
        print("Opening database connection")
        self.conn = "DB Connection"
        return self.conn

    def __exit__(self, exc_type, exc_val, exc_tb):
        print("Closing database connection")
        if exc_type:
            print(f"Error occurred: {exc_val}")
        return False  # Don't suppress exception

# Usage
try:
    with DatabaseConnection() as db:
        print(f"Using {db}")
        raise ValueError("Query error")
except ValueError:
    print("Handling error")
```
