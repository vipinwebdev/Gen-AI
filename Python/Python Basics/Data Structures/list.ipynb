## 1. List
    # A list in Python is an ordered, mutable collection of elements that can store items of different data types and allows duplicate values.
        # Key properties:
            # Ordered ✅
            # Mutable (can change) ✅
            # Allows duplicates ✅
            # Supports multiple data types ✅

```python
my_list = [1, "Hello", 3.5, True]
print("List: ", my_list)

# Indexing
print("Indexing:")
print(my_list[0])
print(my_list[2])
```

```python
# List slicing can help us out with retrieving specific data points from list.
print(my_list[0])
print(my_list[1:3])
my_list[2] = "Welcome"
```

```python
# Reverse
print("Reverse:", my_list[::-1])
my_list.reverse()
print("List: ", my_list)
```

## Python List Methods

```python
# append() -> Adds an element at the end of the list
my_list.append("Add at last.")
print("List: ", my_list)
```

```python
# extend() -> Adds multiple elements (from another iterable)
my_list.extend([1, 2, 3])
print("List: ", my_list)
```

```python
# insert() -> Inserts element at a specific index
my_list.insert(2, "Insert")
print("List: ", my_list)
```

```python
# remove() -> Removes first occurrence of a value
my_list.remove(2)
print("List: ", my_list)
```

```python
# pop() -> Removes and returns element by index (default last)
my_list.pop()
print("List: ", my_list)
```

```python
# index() -> Returns index of first occurrence
print(my_list.index("Hello"))
```

```python
# count() -> Counts occurrences of a value
print(my_list.count(1))
```

```python
# sort() -> Sorts the list in ascending order (in-place)
print(my_list.sort())
```

```python
# reverse() -> Reverses the list (in-place)
my_list.reverse()
```

```python
# copy() -> Returns a shallow copy of the list
new_list = my_list.copy()
```

```python
# len() -> provides length of the invoking list
print(len(new_list))
```

```python
# clear() -> Removes all elements from the list
# my_list.clear()
```

```python
# enumerate() -> Helps with getting the index value along with the actual value in a combination that we stored in list

lst = ["a", "b", "c"]

for index, value in enumerate(lst):
    print(index, value)
```

## Difference:
    # append() → adds single item
    # extend() → adds multiple items

## List Comprehension
    # List comprehension is a short and concise way to create lists using a single line of code.
    # List comprehension is a compact way to create lists using a single line with optional conditions.

```python
# Basic Syntax
# [expression for item in iterable]

# create a list that is storing the values from 0 to 100
number_list = [item for item in range(100)]
print(number_list)

squares = [x*x for x in range(5)]
print(squares)
```

```python
# With Condition

even_numbers = [x for x in range(10) if x % 2 == 0]
print(even_numbers)
```

```python
# With If-Else

result = ["Even" if x % 2 == 0 else "Odd" for x in range(5)]
print(result)
```

```python
# Nested List Comprehension

matrix = [[i*j for j in range(5)] for i in range(5)]
print(matrix)
```
