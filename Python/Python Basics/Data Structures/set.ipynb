## Set
    A set is an unordered, mutable collection of unique elements (no duplicates allowed).
        Key Features
           - Unordered ❌ (no index)
           - No duplicates ❌
           - Mutable (can add/remove) ✅
           - Fast lookup (uses hashing) ⚡

    Set is an unordered collection of unique elements used for fast lookup and removing
    duplicates.

    Syntax:
        my_set = {1, 2, 3, 4}
    Empty set:
        my_set = set()   # NOT {}

```python
# Example
my_set = {1, 2, 2, 3}
print(my_set)
```

## Set Methods

```python
my_set = {1, 2, 3}

# add() -> Adds an element to the set
my_set.add(4)
print("Set:", my_set)
```

```python
# update() -> Adds elements from another iterable
my_set.update([5, 6])
print("Updated Set:", my_set)
```

```python
# remove() -> Removes element, throws error if not present
my_set.remove(2)
print("After remove:", my_set)
```

```python
# discard() -> Removes element safely (no error if not present)
my_set.discard(10)
print("After discard:", my_set)
```

```python
# pop() -> Removes and returns a random element
removed = my_set.pop()
print("Removed:", removed)
print("Set:", my_set)
```

```python
a = {1, 2}
b = {2, 3}

# union() -> Returns combined unique elements
result = a.union(b)
print("Union:", result)
```

```python
# intersection() -> Returns common elements
result = a.intersection(b)
print("Intersection:", result)
```

```python
# difference() -> Elements present in first but not in second
result = a.difference(b)
print("Difference:", result)
```

```python
# symmetric_difference() -> Elements not common in both sets
result = a.symmetric_difference(b)
print("Symmetric Difference:", result)
```

```python
# issubset() -> Checks if set is subset of another
print({1, 2}.issubset({1, 2, 3}))
```

```python
# issuperset() -> Checks if set is superset of another
print({1, 2, 3}.issuperset({1, 2}))
```

```python
# isdisjoint() -> Returns True if no common elements
print({1, 2}.isdisjoint({3, 4}))
```

```python
# clear() -> Removes all elements from set
my_set.clear()
print("After clear:", my_set)
```

## Set Comprehension
    Set comprehension is a concise way to create a set using a single line of code.
    Syntax:
        {expression for item in iterable}

```python
# Create set of squares
squares = {x*x for x in range(5)}
print(squares)
```

```python
# Only even numbers
even_set = {x for x in range(10) if x % 2 == 0}
print(even_set)
```

```python
lst = [1, 2, 2, 3, 4]

# Convert list to set (removes duplicates)
unique = {x for x in lst}
print(unique)
```

```python
# With If-Else
result = {x if x % 2 == 0 else -x for x in range(5)}
print(result)
```
