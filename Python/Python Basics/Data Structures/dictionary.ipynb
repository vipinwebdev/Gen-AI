## Dictionary

    # A dictionary is an unordered, mutable collection of key–value pairs, where keys are unique.
        # Key Points
            * Keys must be immutable (string, number, tuple)
            * Values can be any type
            * Fast lookup (hashing)

```python
# Syntax
my_dict = {
    "name": "Vipin",
    "age": 25,
    "city": "Delhi"
}
```

```python
# Access Values
print(my_dict["name"])        # Vipin
print(my_dict.get("age"))     # 25
```

```python
# Add / Update
my_dict["age"] = 26           # Update
my_dict["country"] = "India"  # Add
print(my_dict)
```

```python
# Remove Elements
my_dict.pop("city")       # Remove by key
del my_dict["age"]        # Delete key
print(my_dict)
```

```python
# Loop Through Dictionary
for key, value in my_dict.items():
    print(key, value)
```

## Common Dictionary Methods

```python
my_dict = {"name": "Vipin", "age": 25}

# get() -> Returns value for key, avoids error if key not found
value = my_dict.get("name")
print("Value:", value)
```

```python
# keys() -> Returns all keys in dictionary
print("Keys:", my_dict.keys())
```

```python
# values() -> Returns all values in dictionary
print("Values:", my_dict.values())
```

```python
# items() -> Returns key-value pairs as tuples
print("Items:", my_dict.items())
```

```python
# update() -> Adds or updates key-value pairs
my_dict.update({"age": 30, "city": "Delhi"})
print("Updated:", my_dict)
```

```python
# pop() -> Removes and returns value of given key
my_dict.pop("age")
print("After pop:", my_dict)
```

```python
# popitem() -> Removes last inserted key-value pair
my_dict.popitem()
print("After popitem:", my_dict)
```

```python
# copy() -> Returns shallow copy of dictionary
new_dict = my_dict.copy()
print("Copy:", new_dict)
```

```python
# setdefault() -> Returns value if key exists, else sets default
my_dict.setdefault("country", "India")
print("After setdefault:", my_dict)
```

```python
# clear() -> Removes all items from dictionary
my_dict.clear()
print("After clear:", my_dict)
```

## Dictionary Comprehension
    # Dictionary comprehension is a compact way to create dictionaries using a single line with
     optional conditions.
    #Syntax
        # {key_expression: value_expression for item in iterable}

```python
# Create dict of numbers and their squares
squares = {x: x*x for x in range(5)}
print(squares)
```

```python
# Create dict of numbers and their squares
squares = {x: x*x for x in range(5)}
print(squares)
```

```python
# Only even numbers
even_squares = {x: x*x for x in range(10) if x % 2 == 0}
print(even_squares)
```

```python
# From Existing Dictionary
data = {"a": 1, "b": 2, "c": 3}

# Multiply values
updated = {k: v*10 for k, v in data.items()}
print(updated)
```

```python
data = {"a": 1, "b": 2}

# Swap keys and values
swapped = {v: k for k, v in data.items()}
print(swapped)
```

```python
# With If-Else
result = {x: "Even" if x % 2 == 0 else "Odd" for x in range(5)}
print(result)
```

# Nested Dictionary
    # A nested dictionary is a dictionary inside another dictionary, used to store structured or
    # hierarchical data.

    # Syntax:
     nested_dict = {
         "key1": {"subkey1": value1, "subkey2": value2},
         "key2": {"subkey1": value3}
     }

```python
students = {
    "101": {"name": "Vipin", "age": 25},
    "102": {"name": "Rahul", "age": 26}
}

print(students["101"]["name"])
print(students)
```

```python
# Update
students["101"]["age"] = 26

# Add new nested key
students["101"]["city"] = "Delhi"
```

```python
# Loop Through Nested Dictionary

for key, value in students.items():
    print("ID:", key)
    for sub_key, sub_val in value.items():
        print(sub_key, sub_val)
```

```python
# Add New Entry
students["103"] = {"name": "Amit", "age": 24}
```

```python
# Always check key existence to avoid errors:
if "101" in students:
    print(students["101"])
```

## Nested Dictionary Comprehension
    Nested dictionary comprehension is used to create a dictionary where values are also
    dictionaries, using a single line of code.
    Syntax:
        {outer_key: {inner_key: value for inner_key in iterable} for outer_key in iterable}

```python
# Simple Nested Dictionary
# Create multiplication table (1 to 3)
table = {i: {j: i*j for j in range(1, 4)} for i in range(1, 4)}
print(table)
```

```python
# Students with Subjects
students = ["Vipin", "Rahul"]
subjects = ["Math", "Science"]

data = {s: {sub: 0 for sub in subjects} for s in students}
print(data)
```

```python
# With Condition
result = {
    i: {j: i*j for j in range(5) if j % 2 == 0}
    for i in range(3)
}
print(result)
```
