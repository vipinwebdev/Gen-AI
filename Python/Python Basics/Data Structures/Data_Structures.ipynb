## Data Structures in Python

## Data Structures in python are being used for storing the data in a structured format so that we can access the data and along with that we can perform all the operations with these structures that we can perform in the case of databases as well as like CRUD( create, read, update,delete ) operations.

    #Types of data structure in Python
        # 1. List
        # 2. Tuple
        # 3. Set
        # 4. Dictionary


## 1. List
    # A list in Python is an ordered, mutable collection of elements that can store items of different data types and allows duplicate values.
        # Key properties:
            # Ordered ✅
            # Mutable (can change) ✅
            # Allows duplicates ✅
            # Supports multiple data types ✅

```python
my_list = [1, "Hello", 3.5, True]
print("List: ", my_list)

# Indexing
print("Indexing:")
print(my_list[0])
print(my_list[2])
```

## 2. Tuple
    #A tuple in Python is an ordered, immutable collection of elements that can store multiple data types and allows duplicate values.
       # Key properties:
           # Ordered ✅
           # Immutable (cannot change) ❌
           # Allows duplicates ✅
           # Supports multiple data types ✅

```python
my_tuple = (1, "Hello", 3.5, True)
print("Tuple: ", my_tuple)
```
