 ## Tuple
    #A tuple in Python is an ordered, immutable collection of elements that can store multiple
     data types and allows duplicate values.
       # Key properties:
           # Ordered ✅
           # Immutable (cannot change) ❌
           # Allows duplicates ✅
           # Supports multiple data types ✅

```python
# Syntax
my_tuple = (1, "A", 3.5)

t = (10, 20, 30)
print(t[1])   # Output: 20
```

## Tuple Methods

```python
# count() -> Counts how many times a value appears
result = my_tuple.count(20)
print("Count:", result)
```

```python
# index() -> Returns index of first matching value
result = my_tuple.index(20)
print("Index:", result)
```

```python
# Packing -> Multiple values stored in a tuple
my_tuple = 1, 2, 3
print(my_tuple)
```

```python
# Unpacking -> Assign tuple values to variables
a, b, c = (10, 20, 30)
print(a, b, c)
```

```python
# Extended Unpacking (*)

# * is used to collect multiple values
a, *b = (1, 2, 3, 4)

print(a)  # 1
print(b)  # [2, 3, 4]
```

```python
# Ignore Values (_)

# _ is used to ignore values
a, _, c = (10, 20, 30)
print(a, c)
```

## Let’s look at real-world use cases of tuple unpacking (useful for interviews + practical coding).

```python
# 1. Swapping Variables (Very Common)
a = 10
b = 20

# Swap without temp variable
a, b = b, a
print(a, b)
```

```python
# 2. Returning Multiple Values from Function
def get_user():
    return "Vipin", 25

name, age = get_user()
print(name, age)
```

```python
# 3. Looping with Index (enumerate)

lst = ["a", "b", "c"]

for i, val in enumerate(lst):
    print(i, val)
```

```python
# 4. Iterating Dictionary
data = {"name": "Vipin", "age": 25}

for key, value in data.items():
    print(key, value)
```

```python
# 5. Ignoring Unwanted Values

a, _, c = (10, 20, 30)
print(a, c)
```

```python
# 6. Handling Variable-Length Data (*)
first, *rest = [1, 2, 3, 4, 5]

print(first)  # 1
print(rest)   # [2, 3, 4, 5]
```

```python
# 7. Splitting Data
name, age = "Vipin,25".split(",")
print(name, age)
```

```python
# 8. Working with Database Rows
row = (101, "Vipin", "Delhi")

id, name, city = row
print(name)
```
