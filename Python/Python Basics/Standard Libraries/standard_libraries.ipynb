## What are Standard Libraries?

    # Python's Standard Library is a collection of built-in modules that come with Python
      installation. It provides ready-to-use functionality for common programming tasks without
      requiring external packages.



## Key Categories of Standard Libraries

        1. File and Directory Access:
            os - Operating system interface
            os.path - Path manipulations
            shutil - High-level file operations
            glob - Unix style pathname pattern expansion
            pathlib - Object-oriented filesystem paths (modern approach)

        2. Data Types and Structures
            collections - Specialized container datatypes (namedtuple, deque, Counter, defaultdict)
            array - Efficient numeric arrays
            enum - Enumeration types
            datetime - Date and time manipulation
            decimal - Decimal fixed-point arithmetic

        3. Text Processing
            re - Regular expressions
            string - String constants and utilities
            difflib - Compare sequences
            textwrap - Text wrapping and filling

        4. Mathematics and Numbers
            math - Mathematical functions
            random - Generate random numbers
            statistics - Statistical calculations
            fractions - Rational numbers

        5. Data Persistence
            sqlite3 - SQLite database interface
            pickle - Python object serialization
            shelve - Persistent dictionary storage
            csv - CSV file reading/writing


        6. Internet and Networking
            socket - Low-level networking
            urllib - URL handling
            json - JSON parsing and generation
            http.client - HTTP protocol client
            smtplib - Email sending

## When to Use Standard Libraries vs External Packages?

        # Use Standard Libraries when:
            The functionality is sufficient for your needs
            You want to minimize dependencies
            You need maximum portability
            You're building simple scripts or tools



    # The Python Standard Library is one of Python's greatest strengths, often described as
    "batteries included" because it provides a comprehensive set of tools for most common programming
    tasks right out of the box.

```python
import random

# random.choice() works with a sequence
choice = random.choice([10, 3, 5, 7, 2])  # Pass a list
# or
choice = random.choice((10, 3, 5, 7, 2))  # Pass a tuple

print(choice)  # Will print one random number from the list
```

```python
import datetime
print(datetime.datetime.now())
print(datetime.datetime.now().day)
print(datetime.datetime.now().month)
print(datetime.datetime.now().year)
```

```python
import re

text = "My email is john@example.com and my phone is 123-456-7890"

# Find email
email_match = re.search(r'\w+@\w+\.\w+', text)
if email_match:
    print(f"Found email: {email_match.group()}")  # john@example.com
    print(f"Position: {email_match.span()}")      # (11, 28)

# Find phone number
phone_match = re.search(r'\d{3}-\d{3}-\d{4}', text)
if phone_match:
    print(f"Found phone: {phone_match.group()}")  # 123-456-7890
```

```python
import re

text = "Apples cost $5, oranges $3, bananas $2"

# Find all prices
prices = re.findall(r'\$\d+', text)
print(prices)  # ['$5', '$3', '$2']

# Find all words
words = re.findall(r'\w+', text)
print(words)  # ['Apples', 'cost', '5', 'oranges', '3', 'bananas', '2']
```
