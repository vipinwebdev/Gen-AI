# Operator Overloading in Python

Operator overloading allows us to define how built-in operators such as `+`, `-`, `==`, and `<` behave for objects of a user-defined class.

Python supports operator overloading through **special methods** (also called magic methods or dunder methods), such as `__add__()` and `__eq__()`.


## Why Use Operator Overloading?

Operator overloading makes user-defined objects behave more naturally.

It is useful when:
- objects represent mathematical values such as vectors, points, or matrices
- objects need readable comparison logic
- objects should print in a cleaner and more meaningful way

Without operator overloading, we would need method calls like `obj1.add(obj2)` instead of simply writing `obj1 + obj2`.


## Common Operators and Their Special Methods

| Operator | Special Method | Meaning |
| --- | --- | --- |
| `+` | `__add__(self, other)` | Addition |
| `-` | `__sub__(self, other)` | Subtraction |
| `*` | `__mul__(self, other)` | Multiplication |
| `/` | `__truediv__(self, other)` | Division |
| `==` | `__eq__(self, other)` | Equality check |
| `<` | `__lt__(self, other)` | Less than |
| `<=` | `__le__(self, other)` | Less than or equal to |
| `>` | `__gt__(self, other)` | Greater than |
| `>=` | `__ge__(self, other)` | Greater than or equal to |
| `str()` | `__str__(self)` | User-friendly string form |
| `repr()` | `__repr__(self)` | Developer-friendly string form |


## Example 1: Overloading `+` with a `Point` Class

In this example, two `Point` objects are added by combining their `x` and `y` coordinates.


````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Point({self.x}, {self.y})"


p1 = Point(2, 3)
p2 = Point(4, 5)
result = p1 + p2

print("First point:", p1)
print("Second point:", p2)
print("After addition:", result)
````

### Explanation

When Python evaluates `p1 + p2`, it internally calls:

```python
p1.__add__(p2)
```

The returned object is another `Point` whose coordinates are the sum of both points.


## Example 2: Overloading Comparison Operators

Here, a `Student` object is compared based on marks.


````python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __gt__(self, other):
        return self.marks > other.marks

    def __eq__(self, other):
        return self.marks == other.marks

    def __str__(self):
        return f"{self.name} ({self.marks})"


s1 = Student("Amit", 85)
s2 = Student("Neha", 92)
s3 = Student("Ravi", 85)

print(s1, ">", s2, ":", s1 > s2)
print(s2, ">", s1, ":", s2 > s1)
print(s1, "==", s3, ":", s1 == s3)
````

## Example 3: Overloading `*` to Scale a Vector

A vector can be multiplied by a number to scale its magnitude.


````python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __mul__(self, scalar):
        return Vector(self.x * scalar, self.y * scalar)

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"


v = Vector(3, 4)
print("Original vector:", v)
print("Scaled vector:", v * 2)
````

## Important Notes

- Operator overloading does not change the meaning of operators globally. It only changes behavior for the class where the method is defined.
- The method should return a meaningful value or object.
- If the operation is not valid, it is better to raise an error or return `NotImplemented`.
- Use operator overloading only when it makes code more readable and intuitive.


## Best Practice Example with Validation

The following version checks the type before performing addition. This is a cleaner and safer implementation.


````python
class SafePoint:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if not isinstance(other, SafePoint):
            return NotImplemented
        return SafePoint(self.x + other.x, self.y + other.y)

    def __repr__(self):
        return f"SafePoint({self.x}, {self.y})"


sp1 = SafePoint(1, 2)
sp2 = SafePoint(5, 7)
print(sp1 + sp2)
````

## Summary

Operator overloading in Python is implemented using special methods.

It helps objects behave like built-in types, which improves readability and expressiveness. Commonly overloaded methods include `__add__`, `__mul__`, `__eq__`, and `__str__`.

When used carefully, operator overloading makes object-oriented programs easier to understand and use.
