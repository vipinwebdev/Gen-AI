## Abstraction in OOP
- Abstraction is one of the important pillars of Object-Oriented Programming.
- It means hiding the internal implementation details and showing only the essential features.
- The user focuses on what an object does, not on how it does it.


- Why abstraction is important:
    - It reduces complexity in the program.
    - It improves readability and maintainability.
    - It allows the user to interact with simple interfaces.
    - It hides unnecessary internal details.


- In Python, abstraction is commonly achieved using:
    - Abstract classes
    - Abstract methods

- An abstract class acts like a blueprint for other classes.
- An abstract method is a method declared in the parent class but implemented in child classes.


- Basic example:
    - If we have a `Vehicle` class with a method `start()`, the user only knows that the vehicle starts.
    - The internal process of starting the engine is hidden from the user.
    - This is called abstraction.


```python
from abc import ABC, abstractmethod


class Vehicle(ABC):
    @abstractmethod
    def start(self):
        print("Vehicle is starting")


class Car(Vehicle):
    def start(self):
        print("Car starts with a key or button")

class Bike(Vehicle):
    def start(self):
        print("Bike starts with a key or kick")


c = Car()
c.start()

bike = Bike()
bike.start()
```

- Summary:
    - Abstraction hides implementation details.
    - It shows only the necessary features to the user.
    - It is useful for reducing complexity and improving code design.


- More examples of abstraction:
    - ATM machine:
        - The user only interacts with options like withdraw, deposit, and check balance.
        - The internal banking process is hidden from the user.

    - Mobile phone:
        - We make calls, send messages, and use apps.
        - The internal hardware and signal processing are hidden.

    - Car:
        - The driver uses steering, brake, and accelerator.
        - The internal engine and mechanical operations remain hidden.


```python
from abc import ABC, abstractmethod


class Shape(ABC):
    @abstractmethod
    def area(self):
        pass


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius


class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def area(self):
        return self.length * self.breadth


c = Circle(5)
r = Rectangle(4, 6)

print(c.area())
print(r.area())
```

```python
from abc import ABC, abstractmethod


class Employee(ABC):
    @abstractmethod
    def salary(self):
        pass


class FullTimeEmployee(Employee):
    def salary(self):
        return "Full-time employee gets monthly salary"


class Freelancer(Employee):
    def salary(self):
        return "Freelancer gets payment per project"


e1 = FullTimeEmployee()
e2 = Freelancer()

print(e1.salary())
print(e2.salary())
```

- Example of abstract class:
    - `Animal` is an abstract class.
    - `sound()` is an abstract method.
    - Child classes like `Dog` and `Cat` provide their own implementation of `sound()`.
    - This helps in defining a common structure for all child classes.


```python
from abc import ABC, abstractmethod


class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass


class Dog(Animal):
    def sound(self):
        return "Bark"


class Cat(Animal):
    def sound(self):
        return "Meow"


d = Dog()
c = Cat()

print(d.sound())
print(c.sound())
```
