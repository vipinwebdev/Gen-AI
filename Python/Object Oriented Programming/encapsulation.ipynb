## Encapsulation in OOP
- Encapsulation is one of the important pillars of Object-Oriented Programming.
- It means wrapping data and methods together inside a single unit called a class.
- It also helps in controlling access to the internal data of an object.


- Features of encapsulation:
    - Data and related methods stay together in one class.
    - Internal object details can be protected from direct access.
    - Access to data is usually given through methods.


- Benefits of encapsulation:
    - Improves data security.
    - Reduces unwanted changes in the program.
    - Makes code easier to manage and maintain.
    - Helps in achieving better modularity.


- In Python, encapsulation is commonly understood through:
    - Public members
    - Protected members
    - Private members


- Scope and access of members:
    - Public members
        - Scope: Public members can be accessed throughout the program.
        - Access: They can be accessed inside the class, outside the class, and also by child classes.
        - These members are created normally without using underscore symbols.

    - Protected members
        - Scope: Protected members are intended to be used within the class and its derived classes.
        - Access: They can be accessed outside the class as well, but it is not recommended by convention.
        - These members are created using a single underscore `_` before the name.

    - Private members
        - Scope: Private members are limited to the class in which they are defined.
        - Access: They cannot be accessed directly outside the class in the usual way.
        - These members are created using double underscores `__` before the name.
        - Python uses name mangling to restrict direct access.


- Basic examples:
    - Public member example:
        - A variable like `student.name` can be accessed directly from outside the class.

    - Protected member example:
        - A variable like `student._roll_no` should be used inside the class or subclass.
        - It can still be accessed outside, but it is not recommended.

    - Private member example:
        - A variable like `student.__marks` is intended to be used only inside the same class.
        - Direct access from outside is restricted because of name mangling.


```python
class Student:
    def __init__(self):
        self.name = "Jerry"          # Public
        self._roll_no = 101          # Protected
        self.__marks = 95            # Private

    def show_private_data(self):
        return self.__marks

    @property
    def roll_no(self):
        return self._roll_no


student = Student()

print(student.name)
print(student.roll_no)
# print(student.marks)  # Private
print(student.show_private_data())
```

- dir(object)

    - In your encapsulation example, it is useful because it lets you inspect what Python stores inside the
      object, including name-mangled private members.
    - You can explain it like this in notes style:

        - `dir(student)`:
            - `dir()` is a built-in Python function.
            - It returns a list of all attributes and methods available in the object.
            - It helps us inspect what members an object contains.

        - In the case of `student` object:
            - Public member like `name` will appear directly.
            - Protected member like `_roll_no` will also appear.
            - Private member like `__marks` will not appear with the same name.
            - It usually appears in name-mangled form such as `_Student__marks`.

        - Why it is useful:
            - It helps in debugging.
            - It helps in understanding object structure.
            - It shows how Python handles private members internally.
