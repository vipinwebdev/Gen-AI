# ========== 1. INHERITANCE ==========
   - Creating class hierarchies for code reuse
   - Child classes inherit from parent classes
   - Single, multiple, and multi-level inheritance
   - super() to call parent methods
   - Promotes code reuse and logical relationships

```python
# Creating new classes based on existing classes, promoting code reuse

# Parent
class Avengers:
    def __init__(self, name, power):
        self.name = name
        self.power = power

    def introduction(self):
        print(f"Hello my name is {self.name} and my power is {self.power}")

# Child
class Ironman(Avengers):
    def __init__(self, name, power, weapon):
        super().__init__(name, power)
        self.weapon = weapon

    def weapon_details(self):
        print(f"I use {self.weapon}")

ironman = Ironman("Tony", "'Rich Genius'", "Suit")

print(ironman.name)
ironman.introduction()
ironman.weapon_details()
ironman.power

```

- Multilevel Inheritance

```python
class Grandparent:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def introduction(self):
        print(f"Hello my name is {self.name} I am a {self.gender} and my age is {self.age}\n")
```

```python
class Parent(Grandparent):
    def __init__(self, name, age, gender,nationality):
        super().__init__(name,age,gender)
        self.nationality = nationality

    def parent_introduction(self):
        print(f"Hello my name is {self.name} I live in {self.nationality} and my age is {self.age}\n")
```

```python
class Child(Parent):
    def __init__(self, name, age, gender,nationality,charm):
        super().__init__(name, age, gender, nationality)
        self.charm = charm

    def child_introduction(self):
        print(f"I you want to know whether I am charming or not then the answer is {self.charm}\n")
```

```python
grand_parent = Grandparent("Sizumi", "63", "Female")
parent = Parent("Uzimaroo", "36", "Male", "Herosima")
child = Child("Nobita", "12", "Male", "Tokyo", "No")

grand_parent.introduction()

parent.introduction()
parent.parent_introduction()

child.introduction()
child.parent_introduction()
child.child_introduction()
```

- Multiple Inheritance

```python
class Ironmans:
    def __init__(self, name, power, suit, weapon, nationality="American"):
        self.name = name
        self.power = power
        self.suit = suit
        self.weapon = weapon
        self.nationality = nationality
    def introduction(self):
        print(f"Hello my name is {self.name} and my power is {self.power} and my suit is {self.suit} and I am {self.nationality}.\n")

class CaptainAmerica:
    def __init__(self, name, power, suit, weapon, nationality="American"):
        self.name = name
        self.power = power
        self.suit = suit
        self.weapon = weapon
        self.nationality = nationality

    def weapon_details(self):
        print(f"I carry the weapon as {self.weapon}")
```

```python
class Hulk(Ironmans, CaptainAmerica):
    def __init__(self, name, power, suit, weapon,anger_issue, nationality="American"):
        super().__init__(name, power, suit,weapon,nationality)
        self.anger_issue = anger_issue

    def anger_description(self):
        print(f"If you want to know whether I have anger issues or not then answer is {self.anger_issue}\n")
```

```python
hulk_guy = Hulk("Hulk", "Amazing Power An strength", "Mark 42", "Repulsor beams and Shield","Yes")

hulk_guy.introduction()
hulk_guy.weapon_details()
hulk_guy.anger_description()
```

- Some examples

```python
class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

    def make_sound(self):
        return "Some generic animal sound"

    def eat(self):
        return f"{self.name} is eating"

    def info(self):
        return f"{self.name} is a {self.species}"

```

```python
# Single Inheritance
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name, species="Dog")  # Call parent constructor
        self.breed = breed

    def make_sound(self):  # Method overriding
        return "Woof! Woof!"

    def fetch(self):  # New method specific to Dog
        return f"{self.name} is fetching the ball"

class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name, species="Cat")
        self.color = color

    def make_sound(self):
        return "Meow!"

    def scratch(self):
        return f"{self.name} is scratching"
```

```python
# Multiple Inheritance
class Flyable:
    def fly(self):
        return "Flying through the air"

class Swimmable:
    def swim(self):
        return "Swimming in water"

class Duck(Animal, Flyable, Swimmable):
    def __init__(self, name):
        super().__init__(name, species="Duck")

    def make_sound(self):
        return "Quack! Quack!"
```

```python
# Demonstrate inheritance
dog = Dog("Max", "Golden Retriever")
cat = Cat("Luna", "Black")
duck = Duck("Donald")

print(dog.info())
print(dog.make_sound())
print(dog.fetch())
print(dog.eat())

print("\n")
print(cat.info())
print(cat.make_sound())
print(cat.scratch())

print("\n")
print(duck.info())
print(duck.make_sound())
print(duck.fly())
print(duck.swim())
```

```python
# Check inheritance hierarchy
print(f"\nIs dog an Animal? {isinstance(dog, Animal)}")
print(f"Is duck a Flyable? {isinstance(duck, Flyable)}")
print(f"Dog class MRO: {[cls.__name__ for cls in Dog.__mro__]}")
```
