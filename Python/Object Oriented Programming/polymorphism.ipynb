========== 2. POLYMORPHISM ==========
   - Same interface, different implementations
   - Method overriding
   - Duck typing
   - Operator overloading (magic methods)
   - Functions/collections that work with multiple types

-  Example 1: Method Overriding (already shown in inheritance)
-  Same method name (make_sound) behaves differently for different classes

-  Example 2: Duck Typing (if it walks like a duck and quacks like a duck...)

```python
class Bird:
    def fly(self):
        return "Bird is flying with wings"

class Airplane:
    def fly(self):
        return "Airplane is flying with engines"

class Superman:
    def fly(self):
        return "Superman is flying with cape"

def make_it_fly(flying_object):
    """Polymorphic function - works with any object that has fly() method"""
    return flying_object.fly()
```

```python
bird = Bird()
plane = Airplane()
superman = Superman()

print(make_it_fly(bird))
print(make_it_fly(plane))
print(make_it_fly(superman))
```

- Example 3: Operator Overloading (Python's magic methods)

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):  # Overload + operator
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):  # Overload - operator
        return Vector(self.x - other.x, self.y - other.y)

    def __str__(self):  # String representation
        return f"Vector({self.x}, {self.y})"

    def __eq__(self, other):  # Overload == operator
        return self.x == other.x and self.y == other.y
```

```python
v1 = Vector(2, 3)
v2 = Vector(4, 5)

print(f"v1: {v1}")
print(f"v2: {v2}")
print(f"v1 + v2: {v1 + v2}")
print(f"v2 - v1: {v2 - v1}")
print(f"v1 == Vector(2, 3): {v1 == Vector(2, 3)}")
```

- Example 4: Different classes with same method name

```python
class Square:
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side ** 2

    def shape_name(self):
        return "Square"

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14159 * self.radius ** 2

    def shape_name(self):
        return "Circle"

class Triangle:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height

    def shape_name(self):
        return "Triangle"
```

```python
# Polymorphic collection
shapes = [Square(4), Circle(3), Triangle(6, 8)]

print("\nCalculating areas polymorphically:")
for shape in shapes:
    print(f"{shape.shape_name()}: Area = {shape.area()}")
```
