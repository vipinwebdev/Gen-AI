## 4 PILLARS OF OBJECT-ORIENTED PROGRAMMING (OOP)
- INHERITANCE
- POLYMORPHISM
- ENCAPSULATION
- ABSTRACTION

- Inheritance
    - Inheritance is a fundamental concept of OOP because of which the child class is able to inherit the
      properties or attributes and functionalities of the parent class.
    - Just like a child is going to inherit the properties from the grandparent and definitely in this
      hierarchy the grand child can have some features different from the parent and grandparent as well.

- In Python typically we get to see 3 types of Inheritance:
    - Single inheritance
    - Multilevel Inheritance
    - Multiple Inheritance


- Single inheritance
    - In single inheritance, one child class inherits from one parent class.
    - This is the simplest form of inheritance.
- Multilevel Inheritance
    -In multilevel inheritance, one class inherits from another class, and then another class inherits from
    it.
    - Example: Grandparent → Parent → Child
- Multiple Inheritance
    - In multiple inheritance, one child class inherits from more than one parent class.
    - This helps in combining features from different parent classes into one child class.

- Polymorphism
    - Polymorphism means one thing can exist in many forms.
    - In programming, the same method name can behave differently for different objects.
    - It allows us to write common code and get different outputs based on the object.
    - Types of polymorphism in Python:
        - Method Overriding
        - Operator Overloading

- Encapsulation
    - Encapsulation means binding the data and methods together into a single unit called class.
    - It is also used to restrict direct access to some variables and methods.
    - This helps in data security and better control over the program.
    - In Python, encapsulation can be seen using:
        - Public members
        - Protected members
        - Private members

- Abstraction
    - Abstraction means hiding the internal implementation details and showing only the essential features
    to the user.
    - The user knows what an object does, but not how it does it internally.
    - This reduces complexity and makes the code easy to use and maintain.
    - In Python, abstraction is commonly achieved using:
        - Abstract classes
        - Abstract methods

- Summary
    - Inheritance helps in reusing the code.
    - Polymorphism helps in performing one task in different ways.
    - Encapsulation helps in wrapping and protecting data.
    - Abstraction helps in hiding unnecessary implementation details.
    - These four pillars together make Object-Oriented Programming powerful and easy to manage.
