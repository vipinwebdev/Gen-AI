## Object-Oriented Programming Concepts.



## 1. Methods.

    # Instance Methods
        - Operate on instance data
        - Have self as first parameter
        ✅ When to Use
            - Use instance methods when:
                - You need to work with object-specific data
                - Behavior depends on individual instance state
        🎯 Use Cases
            - User profile operations (update name, calculate age)
            - Product operations (apply discount to a specific product)
            - Bank account actions (deposit, withdraw)

    # Class Methods
        - Operate on class level
        - Use @classmethod decorator
        - Have cls as first parameter
        ✅ When to Use
            Use class methods when:
                - You need to work with class-level data
                - You want alternative constructors
                - Logic applies to all objects collectively
        🎯 Use Cases
            - Creating objects from different formats (e.g., from JSON, DB, string)
            - Managing shared resources (like counters, configs)
            - Factory methods

    # Static Methods
        - Independent of class and instance
        - Use @staticmethod decorator
        - No self or cls parameters

        ✅ When to Use
            Use static methods when:
                - Logic is related to class conceptually
                - But does not need instance or class data
        🎯 Use Cases
            - Utility/helper functions (validation, calculations)
            - Formatting or transformation logic
            - Common reusable functions grouped inside class


```python
#   Instance Methods

class Student:
    school = "Python High School"  # Class attribute

    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    # Instance method
    def display_info(self):
        return f"{self.name} is in grade {self.grade}"

    def update_grade(self, new_grade):
        self.grade = new_grade
        return f"{self.name}'s grade updated to {self.grade}"
```

```python
#   Class Method

class Student:
    school = "Python High School"
    student_count = 0

    def __init__(self, name, grade):
        self.name = name
        self.grade = grade
        Student.student_count += 1

    # Class method
    @classmethod
    def get_school_name(cls):
        return cls.school

    @classmethod
    def change_school(cls, new_school):
        cls.school = new_school
        return f"School changed to {new_school}"

    @classmethod
    def create_from_string(cls, student_string):
        name, grade = student_string.split('-')
        return cls(name, int(grade))

# Usage
print(Student.get_school_name())           # Python High School
Student.change_school("Advanced Academy")
print(Student.school)                      # Advanced Academy

# Alternative constructor
student = Student.create_from_string("Alice-10")
print(student.name)  # Alice
```

```python
#   Static Methods

class MathOperations:
    @staticmethod
    def add(x, y):
        return x + y

    @staticmethod
    def multiply(x, y):
        return x * y

    @staticmethod
    def is_even(number):
        return number % 2 == 0

# Called directly on class (no instance needed)
print(MathOperations.add(5, 3))      # 8
print(MathOperations.multiply(4, 7)) # 28
print(MathOperations.is_even(10))    # True
```

## 2. Types Of Constructor
        - Default Constructor
        - Paramaterized Constructor


## Default Constructor
        It doesn't have the parameters inside it and it is only having the self keyword.
        It assigns default values to all of the objects.


```python
class Dog:
    def __init__(self):
        self.name = "German"
        self.color = "Brown"
    def show_info(self):
        print(f"{self.name} is in color {self.color}")

dog1 = Dog()
dog1.show_info()
dog2 = Dog()
dog2.show_info()
```

## Parameterized constructor
        It consists of parameters insinde it so that every object can have a specific attribute
        associated with it.
        It allows us to pass arguments while creating an object.

```python
class Dog:
    def __init__(self, bread, color):
        self.bread = bread
        self.color = color
    def show_info(self):
        print(f"{self.bread} is in color {self.color}")
dog1 = Dog("German", "Brown")
dog1.show_info()
dog2 = Dog("Pitbull","Black")
dog2.show_info()
```
