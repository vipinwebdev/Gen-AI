##  Introduction
        # Object-Oriented Programming is a programming paradigm that organizes code around objects rather than functions and logic.
          Objects contain data (attributes) and behaviors (methods).

        # Core Concepts:
            - Classes - Blueprints for creating objects
            - Objects - Instances of classes
            - Encapsulation - Bundling data and methods
            - Inheritance - Creating new classes from existing ones
            - Polymorphism - Same interface, different implementations
            - Abstraction - Hiding complex implementation details

## Basic Class Definition

            my_dog = Dog()

            print(type(my_dog))  # <class '__main__.Dog'>
            print(isinstance(my_dog, Dog))  # True

##  Class with Constructor and Attributes

        class Dog:
            # Class attribute (shared by all instances)
            species = "Canis familiaris"

            # Constructor method (initializes instance attributes)
            def __init__(self, name, age):
                # Instance attributes
                self.name = name
                self.age = age

            # Instance method
            def description(self):
                return f"{self.name} is {self.age} years old"

            def bark(self, sound="Woof"):
                return f"{self.name} says {sound}"

        # Creating instances
        buddy = Dog("Buddy", 3)
        max = Dog("Max", 5)


        # Accessing attributes
        print(buddy.name)      # Buddy
        print(buddy.age)       # 3
        print(buddy.species)   # Canis familiaris

        # Calling methods
        print(buddy.description())  # Buddy is 3 years old
        print(buddy.bark())         # Buddy says Woof
        print(max.bark("Arf!"))     # Max says Arf!


```python
## 1. Example
class Avenger:
    def __init__(self, name,power, weapon, nationality="American"):
        self.name = name
        self.power = power
        self.weapon = weapon
        self.nationality = nationality
    def introduction(self):
        print(f"Hello my name is {self.name} and I am {self.nationality}.")
    def carrying_weapon(self):
        print(f"I am carrying the weapon {self.weapon}.")
    def carrying_power(self):
        print(f"I am carrying power {self.power}.")
```

```python
Iron_Man = Avenger("Iron Man", "Genius Intellect and Technology", "Repulsor Beams")
Iron_Man.introduction()
Iron_Man.carrying_weapon()
Iron_Man.carrying_power()
```

```python
Captain_America = Avenger("Steve Rogers", "Super Strength and Agility", "Shield")
Captain_America.introduction()
Captain_America.carrying_weapon()
Captain_America.carrying_power()
```

```python
Thor = Avenger("Thor", "Thunder Power", "Hammer","Asgardian")
Thor.introduction()
Thor.carrying_weapon()
Thor.carrying_power()
```
