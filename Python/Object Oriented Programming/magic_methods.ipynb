## Magic Methods in Python
- Magic methods are special methods whose names start and end with double underscores.
- They are also called dunder methods.
- These methods allow user-defined objects to behave like built-in Python objects.
- Python calls them automatically for specific operations.


- Why magic methods are important:
    - They make custom classes more natural to use.
    - They help objects work with built-in functions.
    - They support operator overloading.
    - They improve readability and flexibility in object-oriented programming.


- Common magic methods:
    - `__init__` : Used to initialize object attributes.
    - `__str__` : Returns a readable string for users.
    - `__repr__` : Returns a detailed string for developers.
    - `__len__` : Defines behavior for `len()`.
    - `__add__` : Defines behavior for `+`.
    - `__eq__` : Defines behavior for `==`.
    - `__getitem__` : Defines behavior for indexing.
    - `__call__` : Allows an object to be called like a function.


- `__init__`, `__str__`, and `__repr__`:
    - `__init__` runs automatically when an object is created.
    - `__str__` is used when the object is printed using `print()`.
    - `__repr__` gives a more formal representation of the object.
    - These methods help in creating and displaying objects properly.


```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __str__(self):
        return f"Student Name: {self.name}, Marks: {self.marks}"

    def __repr__(self):
        return f"Student('{self.name}', {self.marks})"


student = Student("Vipin", 95)
print(student)
print(repr(student))
```

- `__len__` method:
    - `__len__` is called when we use `len(object)`.
    - It tells Python what value should be returned as the length of the object.
    - It is useful when the object stores a collection of items.


```python
class BookCollection:
    def __init__(self, books):
        self.books = books

    def __len__(self):
        return len(self.books)


library = BookCollection(["Python", "Java", "C++"])
print(len(library))
```

- `__add__` and `__eq__` methods:
    - `__add__` is used when two objects are added using `+`.
    - `__eq__` is used when two objects are compared using `==`.
    - These methods help us define custom behavior for operators.
    - This concept is also called operator overloading.


```python
class Number:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return Number(self.value + other.value)

    def __eq__(self, other):
        return self.value == other.value

    def __str__(self):
        return str(self.value)


n1 = Number(10)
n2 = Number(20)
n3 = Number(10)

print(n1 + n2)
print(n1 == n3)
```

- `__getitem__` method:
    - `__getitem__` is used when we access an object with square brackets.
    - It makes a custom object behave like a list, tuple, or dictionary.
    - Python calls this method for indexing operations.


```python
class MyData:
    def __init__(self, items):
        self.items = items

    def __getitem__(self, index):
        return self.items[index]


data = MyData([100, 200, 300])
print(data[1])
```

- `__call__` method:
    - `__call__` allows an object to behave like a function.
    - When we write `obj()`, Python checks whether the object has a `__call__` method.
    - This is useful in callable classes, decorators, and configurable processors.


```python
class Greeting:
    def __call__(self, name):
        return f"Hello, {name}!"


greet = Greeting()
print(greet("Vipin"))
```

- Important note:
    - We normally do not call magic methods directly.
    - Python calls them automatically when related operations happen.
    - For example, `print(obj)` uses `__str__`, `len(obj)` uses `__len__`, and `obj1 + obj2` uses `__add__`.
    - This makes custom objects feel natural to use in Python programs.


- Summary:
    - Magic methods define how objects behave in special situations.
    - They are useful for initialization, printing, comparison, indexing, and operator overloading.
    - They help custom classes integrate well with Python syntax and built-in functions.
